## <font style="color:rgb(50,120,229)">Histogramas de cores</font>     
     
   
A estas alturos do curso, debes saber o que é un histograma, concretamente un histograma de cor. Pero repasarémolos brevemente por se acaso. En pocas palabras, un histograma de cor conta o número de veces que unha determinada intensidade de píxeles (ou rango de intensidades de píxeles) ocorre nunha imaxe. Usando un histograma de cor podemos expresar a distribución real ou **cantidad** de cada cor nunha imaxe. Os recontos para cada intervalo de cor/cor úsanse entón como o noso vector de características.

Como descritores de imaxes, os histogramas de cores tenden a ser bastante potentes. Se decidimos utilizar un histograma de cores **3D** con 8 caixas (bins) por canle, poderiamos representar calquera imaxe de calquera tamaño usando só **8 x 8 x 8 = 512 caixas**, ou un vector de características de **512-D**. O tamaño dunha imaxe non ten ningún efecto no noso histograma de cor de saída, aínda que é conveniente cambiar o tamaño das imaxes grandes a unha dimensión máis manexable para aumentar a velocidade do cálculo do histograma.

Creo que agora é mellor ir a un exemplo e ver exactamente como podemos extraer histogramas de cores das imaxes e usalos como vectores de características.

## <font style="color:rgb(50,120,229)">Clusterin k-medias</font>
K-means é un algoritmo de agrupación onde o seu obxectivo é particionar **n puntos** de datos en **k clusters**. Cada un dos n puntos de datos asignarase ao clúster (coa media) máis próxima. A media de cada grupo chámase o seu "centroide" ou "centro":


Un exemplo de k-means que separan datos 2-d en 3 grupos distintos.

En xeral, a aplicación de k-medias obtén k grupos separados dos n puntos de datos orixinais. Os puntos de datos dentro dun clúster particular considéranse "máis similares" entre si que os puntos de datos que pertencen a outros clústeres.

Nesta exemplo, agruparemos os histogramas de cores extraídos das imaxes do noso conxunto de datos, pero en realidade, podes agrupar calquera tipo de vector de características.

Os histogramas que pertencen a un grupo dado serán máis similares na distribución de cores que os histogramas pertencentes a un grupo separado.

Unha advertencia de k-means é que **necesitamos especificar o número de clusters** que queremos xerar antes inicialmente. Hai algoritmos que seleccionan automaticamente o valor óptimo de k, pero estes algoritmos non os empregaremos neste curso. Polo momento, proporcionaremos manualmente un valor de k=2 para separar as dúas clases de imaxes.

## <font style="color:rgb(50,120,229)">Vexamos estas ideas en acción ...</font>

Primeiro imos definir o noso descritor de imaxe e crearemos unha clase para el (poderiamos meter todos os descritores deste tema dentro dela)chamado LabHistogram para extraer histogramas de cores de imaxes no espazo de cor L*a*b*:

```python
# importamos paquetes
import cv2

class LabHistogram:
	def __init__(self, bins):
		# Almacenamos os bins que queremos para o histograma
		self.bins = bins

	def describe(self, image, mask=None):
		# Pasamos ao espazo de cor L*a*b*, achamos o histograma e normalizamolo
		lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
		hist = cv2.calcHist([lab], [0, 1, 2], mask, self.bins,
			[0, 256, 0, 256, 0, 256])

		# Histograma para versión de OpenCV 3+
		hist = cv2.normalize(hist,hist).flatten()

		# devolvemos o histograma
		return hist
```

A nosa clase, LabHistogram, é inicializada cun só un argumento necesario: o número de bins (caixas) no noso histograma 3D.
O método describe() que acepta dous argumentos:  a imaxe da queremos extraer características e **unha máscara opcional, que nos permitiría extraer características só das partes dunha imaxe que nos interesan**. Se empregas esta opción, terías un descriptor local.

O primeiro paso é converter a nosa imaxe do espazo de cor RGB ao espazo de cor L*a*b*. É moi importante que realicemos o paso de normalización. Se non o fixeramos, entón as imaxes co mesmo contido pero diferentes tamaños terían histogramas dramáticamente diferentes. Pola contra, ao normalizar o noso histograma asegurámonos de que o ancho e o alto da nosa imaxe de entrada **non teñan ningún efecto** no histograma de saída. Finalmente, devolve o noso vector de características do histograma.

Por que estamos a usar o espazo de cor L*a*b*? a razón é a que distancia euclidiana entre dúas cores en L*a*b* ten un significado **perceptivo e perceptible**. E dado que o algoritmo de agrupación de k-means asume un espazo euclidiano, obteremos mellores clusters usando o espazo de cor L*a*b* que RGB ou HSV.


```python
# Importacions estandar
from sklearn.cluster import KMeans
from imutils import paths  #pip install imutils
import numpy as np
import cv2
import numpy as np;
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12.0, 12.0)
matplotlib.rcParams['image.cmap'] = 'gray'
```

```python
# argumentos de onde estan os datos e o numero de clusters
args={'dataset':'./dataset/', 'clusters':2}

# Creamos o obxecto que describe a imaxe por un histograma
# de cor e fixamos o numero de bis desexado
desc = LabHistogram([8, 8, 8])
data = []

# almacenamos os path das imaxes do directorio dataset
imagePaths = list(paths.list_images(args["dataset"]))
imagePaths = np.array(sorted(imagePaths))

# lazo sobre todas as imaxes do dataser
for imagePath in imagePaths:
    # cargamos a imaxe, describimola e metemola na lista
    image = cv2.imread(imagePath)
    hist = desc.describe(image)
    data.append(hist)

# Facemos un clustering sobre os histogramas de cor
clt = KMeans(n_clusters=args["clusters"], random_state=42)
labels = clt.fit_predict(data)

# lazo sobre as equiquetas unicas do array e ordenadas (np.unique)
for label in np.unique(labels):
    #almacenamos os path de todas as imaxes que son asocuiadas coa ultima etiqueta
    labelPaths = imagePaths[np.where(labels == label)]
    
    # lazo sobre todas as imaxes que pertencen a ultima etiqueta
    plt.figure(figsize=[15,15])
    for (i, path) in enumerate(labelPaths):
        # cargamos a imaxe as as visualizamos
        image = cv2.imread(path)
        plt.subplot(1,5,i+1);plt.imshow(image);plt.title("Cluster {}, Imaxe #{}".format(label + 1, i + 1));
```
