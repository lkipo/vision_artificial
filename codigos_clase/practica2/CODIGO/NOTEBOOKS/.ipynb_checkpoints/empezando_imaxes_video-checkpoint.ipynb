# <font style="color:rgb(50,120,229)">Lemos a imaxe</font>
OpenCV permite ler distintos tipos de imaxes (JPG, PNG, etc). Podemos cargar imaxes de gris, color e con canle de transparencia. Empregamos a instrucción [**`imread`**](https://docs.opencv.org/4.1.0/d4/da8/group__imgcodecs.html#ga288b8b3da0892bd651fce07b3bbd3a56) que ten a seguinte sintaxe:

### <font style="color:rgb(8,133,37)">Sintaxe da función </font>
``` python
retval	=	cv2.imread(	filename[, flags]	)
```

Ten **2 argumentos**:

1. `retval` é a imaxe cargada se todo vai ben. En caso contrario é `None`. Isto pasará se o path ou nome son incorrecto ou a imaxe está corrupta.
2. `Path ao ficheiro imaxe`: pode ser un path **absoluto** ou **relativo**. Este é un argumento **obrigatorio**.
3. `Flags`: Estes falgs son empregado para ler unha imaxe nnun xeito particular (por exemplo escala de grises/cor/cun canle alfa). Este é un argumento **opcional** con valor por defecto `cv2.IMREAD_COLOR` ou `1` o cal carga a imaxe como imaxe en cor.

Antes de proceder con exemplo, vexamos cales son os `flags` dispoñibles.

**Flags**
1. **`cv2.IMREAD_GRAYSCALE`** ou **`0`**: Carga a imaxe en escala de grises
2. **`cv2.IMREAD_COLOR`** ou **`1`**: carga a imaxe como de cor. Calquera canle de transparencia sere desbotado. É o valor por defecto.
3. **`cv2.IMREAD_UNCHANGED`** ou **`-1`**: Carga a imaxe coa canle de transparencia (alfa).

# <h1 style="font-size:30px;">2.- Coordenadas da imaxe</h1>

Unha vez que os datos da imaxe se almacenan nunha matriz NumPy 2D, podemos visualizalos empregando a función matplotlib `imshow ()`como veremos máis adiante pero hai de cousas que hai que ter en conta sobre a imaxe mostrada:

-O sistema de coordenadas usado para representar os datos dos píxeles da imaxe ten a orixe na esquina superior esquerda da imaxe. O ancho da imaxe mídese ao longo do eixe x e a altura ao longo do eixe y. Cando estes datos se almacenan nunha matriz 2D, a primeira posición do índice que representa as filas da matriz corresponde ao eixe y e a segunda posición corresponde ao eixe x. Este sistema de coordenadas mantén a convención de que o eixo x representa a dirección horizontal e o eixo y a dirección vertical, pero ao acceder aos datos nunha matriz numPy o primeiro índice representa a coordenada `y` e o segundo índice representa a coordenada` x `. Esta indexación aplícase especificamente ás matrices 2D, pero debes ter en conta que moitas funcións en OpenCV (e outras bibliotecas) poden referirse a unha coordenada de píxeles na notación matemática estándar `(x, y)`. Por exemplo, como veremos en futuros módulos, moitas funcións en OpenCV requiren que se especifique unha coordenada de píxel como `punto (x, y)`. Resumindo, cando se accede a coordenadas de píxeles nunha matriz 2P NumPy a indexación sempre é `[y, x]`, pero hai outras veces nas que as coordenadas de imaxe se especifican como `(x, y)`. 



### <font style = "color:rgb(200,0,0)">NOTA</font>
Mantén en mente que en OpenCV, **size** representa unha tupla de `ancho X alto` ou `#columnas X #filas`. Pero en numpy, o método **shape** retorna o tamaño como unha tupla de `altura X anchura`.

### <font style = "color:rgb(200,0,0)">NOTA sobre o indexado (a pesar de correr o risco de ser pesado!)</font>

Como mencionamos anteriormente, as matrices son array de numpy, o primeiro índice será o `número de fila` e o segundo será `número de columna`. Isto é unha fonte de confusión permanente dado que o natural sería pensar nos pixel en termos de coordenadas `(x,y)` ou `(columna,fila)` non en `(fila,columna)`

Por exemplo, para acceder ao elemento na `4ª fila` e na `3ª columna`, debemos empregar `img[3,4]`. Pero como veremos na sección de anotación sobre imaxes, trataremos con puntos que son representados como `(x,y)` e entón, as coordenadas serán `(4,3)`.

# 5. Redimensionar imaxes e volteo de imaxes


<hr style = "border: none; height: 4px; background-color: # D3D3D3" />

A función **`resize ()`** redimensiona a imaxe `src` ata o tamaño especificado. O tamaño e o tipo derívanse de "src", "dsize", "fx" e "fy".

### <font color = "green"> Sintaxe de funcións </font>

```Python
dst = cv2.resize (src, dsize [, dst [, fx [, fy [, interpolación]]]])
```

`dst`: imaxe de saída; ten o tamaño `dsize` (cando non é cero) ou o tamaño calculado a partir de `src.size ()`,`fx` e `fy`; o tipo de `dst` é o mesmo que de` src`.

A función ten **2 argumentos necesarios**:
    
1. `src`: imaxe de entrada
2. `dsize`: tamaño da imaxe de saída

Os argumentos opcionais que se usan a miúdo inclúen:

1. `fx`: factor de escala ao longo do eixo horizontal
2. `fy`: factor de escala ao longo do eixo vertical

A imaxe de saída ten o tamaño `dsize` (cando non é cero) ou o tamaño calculado a partir de` src.size () `,` fx` e `fy`; o tipo de `dst` é o mesmo que de` src`. Se se especifican `fx` e `fy`, entón `dsize` debería establecerse en `None` xa que é un argumento obrigatorio.

The function **`resize()`** resizes the image `src` down to or up to the specified size. The size and type are derived from the `src`, `dsize`, `fx`, and `fy`.


### <font color="green">ocumentación de OpenCV</font>

[**`resize()`**](https://docs.opencv.org/4.5.2/da/d54/group__imgproc__transform.html#ga47a974309e9102f5f08231edc7e7529d)

<hr style="border:none; height: 4px; background-color:#D3D3D3" />

````python
img_cor = cv2.imread('../data/images/checkerboard_color.png', cv2.IMREAD_COLOR)

# Método 1: Redimencionamos a imaxe.
resized_image_2x = cv2.resize(img_cor, None, fx = 2, fy = 2)

# Visualizamos as imaxes.
plt.figure(figsize = [20,5])
plt.subplot(121); plt.imshow(img_cor); plt.title('Orixinal')
plt.subplot(122); plt.imshow(resized_image_2x[:, :, ::-1]); plt.title('Orixinal x 2')
````

````python
# Método 2: Redimencionamos a imaxe a un tamaño desexado.
desired_width  = 400
desired_height = 200
dim = (desired_width, desired_height)

# Redimensionamos.
resized_image = cv2.resize(img_cor, dsize = dim, interpolation = cv2.INTER_AREA)

plt.figure(figsize = [20,5])
plt.subplot(121); plt.imshow(img_cor); plt.title('Orixinal')
plt.subplot(122); plt.imshow(resized_image [:, :, ::-1]); plt.title('Orixinal x 2')
````

# 5.1. Voltear imaxes

<hr style = "border: none; height: 4px; background-color: # D3D3D3" />

A función **`flip ()`** cambia a matriz dun destes tres xeitos diferentes (os índices de fila e columna están baseados en 0):

### <font color = "green"> Sintaxe de funcións </font>
```Python
dst = cv2.flip (src, flipCode [, dst])
```

`dst`: matriz de saída do mesmo tamaño e tipo que src.

A función ten **2 argumentos necesarios**:
    
1. `src`: imaxe de entrada
2. `flipCode`: unha bandeira para especificar como virar a matriz; 0 significa dar a volta ao eixo x e o valor positivo (por exemplo, 1) significa dar a volta ao eixo y. O valor negativo (por exemplo, -1) significa dar a volta a ambos os eixos.


### <font color = "green"> Documentación OpenCV </font> 

[**`flip()`**](https://docs.opencv.org/4.5.2/d2/de8/group__core__array.html#gaca7be533e3dac7feb70fc60635adf441)

<hr style="border:none; height: 4px; background-color:#D3D3D3" />


````python
# Volteamos a imaxe das tres maneiras
img_barco = cv2.imread('../data/images/barco.jpg', cv2.IMREAD_COLOR)
img_barco_flipped_horz = cv2.flip(img_barco , 1)
img_barco_flipped_vert = cv2.flip(img_barco , 0)
img_barco_flipped_both = cv2.flip(img_barco , -1)

# isualizamos as imaxes.
plt.figure(figsize = [18, 5])
plt.subplot(141); plt.imshow(img_barco_flipped_horz[:, :, ::-1]); 
plt.title('Volteo horizontal');
plt.subplot(142); plt.imshow(img_barco_flipped_vert[:, :, ::-1]); 
plt.title('Volteo  Vertical');
plt.subplot(143); plt.imshow(img_barco_flipped_both[:, :, ::-1]); 
plt.title('Volteo dobre');
plt.subplot(144); plt.imshow(img_barco[:, :, ::-1]);              
plt.title('Orixinal');
````

# <font style="color:rgb(50,120,229)">6. Propiedades dunha imaxe: cor e canles</font>

Como mencionamos anteriormente, unha imaxe de cor consiste en múltiples canles e cada unha delas é coma unha imaxe de gris. **`A combinación de valores de intensidade das tres canles dunha imaxe de cor da como resultado a imaxe amosada sobre a pantalla do ordenador`**. Hai vairos espazos de cor que se empregan en procesado de imaxes pero para o almacenamento, o máis popular, é o espazo RGB. 

En OpenCV, a orden das canles R, G e B está **invertida**. i.e. na matriz imaxe, a canle azul (B) está indexada primeiro, seguida pola canle verde (G) e finalmente pola vermella (R). O resultado é o espazo BGR e non RGB (non hai razón obxectivas para isto só éun criterio)

### <font style = "color:rgb(8,133,37)">Visualizando a imaxe</font>

````python
# Visualización da imaxe
plt.imshow(img)
````

#### <font style = "color:rgb(200,0,0)">Importante!</font>

Notas algo raro na imaxe de cor visialiada aqui? Isto é porque OpenCV emprega formato **BGR** por defecto mentres que Matplotlib asume que a imaxe será **RGB**. Isto pode ser correxido se: 

* convertemos a imaxe ao espazpo RGB empregando **`cv2.cvtColor(img, cv2.COLOR_BGR2RGB)`** ou 
* invertindo a orden das canles - **`plt.imshow(img[:,:,::-1])`** intercambia a orden das 1ª e 3ª canle.

````python
# Convertemos o espazo de cor BGR a RGB 
imgRGB = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(imgRGB)
````

````python
# Tamén podemos empregar a seguinte instrucción - Invertirá a orden das canles BGR a RGB
plt.imshow(img[:,:,::-1])
````

Acedemos ás tres canles e visualizamos a imaxe de gris para cada canle

````python
# Visualizamos as canles
plt.figure(figsize=[20,5])

plt.subplot(131);plt.imshow(img[:,:,0]);plt.title("Canle B (Azul)")
plt.subplot(132);plt.imshow(img[:,:,1]);plt.title("Canle G (Verde)")
plt.subplot(133);plt.imshow(img[:,:,2]);plt.title("Canle R (Vermella)")
````

#### <font style = "color:rgb(8,133,37)">Observación </font>
Como xa mencionamos, un pixel branco ten un valor alto na matriz de intensidades (255).Se observamos as canles detidadamente e o comparamos coa imaxe orixinal, veremos que:

1. na imaxe orixinal o fondo é azul. Entón, a dimensión azul debe ter valores máis elevados que o resto de canles (próximos a cero) nos puntos do fondo.
2. a cara é cor carne (sonsorrada), polo tanto, para estes pixels a canle de cor vermella ten que ter valores altos fronte ás outras dimensións, que serán máis baixos.
3. a imaxe ten un tono verdoso dado que a canle G ten valores altos en todos os puntos da imaxe.

Este tipo de información pode ser moi útil se queres construir un aplicación básica que tome decisións baseadas na cor (ou máis ainda, empregando unha canle específica de cor)


# División e fusión de canles de cor

Aprendamos agora a dividir e combinar canles de cor usando as funcións ** `split ()` ** e ** `merge ()` ** en OpenCV.

<hr style = "border: none; height: 4px; background-color: # D3D3D3" />

** `split ()` ** Divide unha matriz multicanle en varias matrices de canle único.

** `merge ()` ** Combina varias matrices para facer unha única matriz multicanle. Todas as matrices de entrada deben ter o mesmo tamaño. 

Vexamos como se pode facer isto.

### <font color="green">Documentación en OpenCV</font>

[**`split()`**](https://docs.opencv.org/4.5.2/d2/de8/group__core__array.html#ga0547c7fed86152d7e9d0096029c8518a)
[**`merge()`**](https://docs.opencv.org/4.5.2/d2/de8/group__core__array.html#ga61f2f2bde4a0a0154b2333ea504fab1d)

<hr style="border:none; height: 4px; background-color:#D3D3D3" />

````python
# Separa a as canles B,G,R da imaxe
b,g,r = cv2.split(img)

# Visualiza as canles
plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(b);plt.title("Canle B")
plt.subplot(142);plt.imshow(g);plt.title("Canle G")
plt.subplot(143);plt.imshow(r);plt.title("Canle R")

# Unimos as canles BGR da imaxe
imgMerged = cv2.merge((b,g,r))

# Visualizamos a unión
plt.subplot(144);plt.imshow(imgMerged[:,:,::-1]);plt.title("Canles unidas");
````

# <font style="color:rgb(50,120,229)">7. Captura, visualización e escritura de vídeo</font>
Un video é unha secuencia de imaxes que vai variando co tempo. O xeito de almacenalos é nun array de catro dimensións (R,G,B,t). Igual que cando tratabamos con imaxes e empregamos a función [**`cv2.imread`**](https://docs.opencv.org/4.1.0/d4/da8/group__imgcodecs.html#ga288b8b3da0892bd651fce07b3bbd3a56) de OpenCV para ler unha imaxe almacenada no noso disco duro, esta libraría tamén nos aporta unha función  **`cv2.VideoCapture`** para criar un [**VideoCapture**](https://docs.opencv.org/4.1.0/d8/dfe/classcv_1_1VideoCapture.html#ac4107fb146a762454a8a87715d9b7c96) que crea un obxecto que nos permite ler un vídeo dende un ficheiro indicado.

### <font style="color:rgb(8,133,37)">Sintaxe da función </font>

``` python
<VideoCapture object>	=	cv2.VideoCapture(		)
<VideoCapture object>	=	cv2.VideoCapture(	filename[, apiPreference]	)
<VideoCapture object>	=	cv2.VideoCapture(	index[, apiPreference]	)

```

**Parámetros**

- **`filename`** pode ser:
    - nome dun ficheiro de vídeo (ie. video.avi)
    - ou unha secuencia de imaxes (ie. img_%02d.jpg, lera todos os ficheiros co nomes como img_00.jpg, img_01.jpg, img_02.jpg, ...)
    -ou URL do video stream (ie. protocolo://host:port/script_name?script_params|auth). Nota que cada video stream ou camara IP ten o seu propio esquema URL. Consulta a documentación correspondente para introducir a URL correcta.
- **`apiPreference`**:	API preferida de captura que queremos empregar.Pode empregar para forzar o uso dun lector específico se hai varios dispoñibles, por exemplo: cv::CAP_FFMPEG ou cv::CAP_IMAGES ou cv::CAP_DSHOW.

## <font style="color:rgb(50,120,229)">Creando un obxecto para ler un video </font>
```python
cap = cv2.VideoCapture(args)
```

As tres formas máis populares de ler videos empregando o obxecto VideoCapture son:
1. Empregando a Webcam ( pasamos o argumento 0 )
2. Debde un ficheiro de video ( especificamos o nome do ficheiro )
3. Secuencia de imaxes ( especificamos o patrón da secuencia image_%03d.jpg )

#### <font style="color:rgb(200,0,0)">NOTE para Webcam</font>

En moitos casos, só unha cámara está conectada ao noso sistema polo que só temos que pasar un argumento **`0`** e OpenCV emprega a única cámara ligada ao noso ordenador. Cando temos máis dunha cámara conectada ao noso sistema, podemos seleccionar unha segunda cámara pasando o argumento  **`1`**, a terceira cámara pasando o argumento **`2`** e así sucesivamente.

## <font style="color:rgb(50,120,229)">Visualizano os frames como un video [ para local só! ]</font>

Como no caso dunha imaxe, empregaremos [**`cv2.waitKey()`**](https://docs.opencv.org/4.1.0/d7/dfc/group__highgui.html#ga5628525ad33f52eab17feebcfba38bd7) despois da función [**`cv2.imshow()`**](https://docs.opencv.org/4.1.0/d7/dfc/group__highgui.html#ga453d42fe4cb60e5723281a89973ee563) para pausar cada frame de vídeo. No caso dunha imaxe, pasamos un  **`0`** á función `waitKey`, posto que pausaría un tempo indefinido a visualización dun frame. Como un vídeo debe dar sensación de continuidade, precisamos pasar un número razonablemente maior que  `0` a waitKey(). **Este número é o tempo, en milisegundos, que desexamos visualizar cada frame.**

#### <font style="color:rgb(50,120,229)">Valor de waitKey para Webcam</font>
Cando estamos lendo frame dende a webcam, empregaremos `waitKey(1)` porque así o frame rate da visualización estará limitado polo frame rate da cámara.

#### <font style="color:rgb(50,120,229)">Valor de waitKey para un ficheiro de video</font>
Mentre que estamos lendo un vído que estamos a procesar, pode ser apropiado asignar un retardo de 1 ms a hebra correspondente e permitir o procesado desesado.

**Escribe un script  `videoRead.py` e comproba o funcionamento deste parámetros.**

````python
```python
# Completamos a lectura do video
while(cap.isOpened()):
  # Capturamos frame-a-frame
  ret, frame = cap.read()
    
  if ret == True:
    cv2.imshow("Video", frame)    
    # Esperamos 25 ms antes de movernos ao seguinte frame
    # Isto enlentecera a visualizacion do video
    cv2.waitKey(25)
    
  # Rompemos o lazo
  else: 
    break
```
````

# <font style="color:rgb(50,120,229)">Consultar e asignar as propiedades de VideoCapture</font>

A continuación veremos como extraer algunha propiedade do vídeo e como cambialas empregando  **`cap.get(propId)`** e **`cap.set(propId,value)`** respectivamente. 

Aqui, **`cap`** é o obxecto **`VideoCapture`** de onde queremos extraer (ou asignar) as propiedades, identificadas por **`propId`** [Property ID](https://docs.opencv.org/4.1.0/d4/d15/group__videoio__flags__base.html#gaeb8dd9c89c10a5c63c139bf7c4f5704d) e **`value`** é o valor que queremos asignarlle á propiedade identificada con id dado por **`propId`**.

Este cadro resume as propiedades máis habituais ID:

| Enumerator | Numerical Value | Property |
| --- | --- | --- |
| **`cv2.CAP_PROP_POS_MSEC`** | `0` | Current position of the video file in milliseconds |
| **`cv2.CAP_PROP_FRAME_WIDTH`** | `3` | Width of the frames in the video stream |
| **`cv2.CAP_PROP_FRAME_HEIGHT`** | `4` | Height of the frames in the video stream |
| **`cv2.CAP_PROP_FPS`** | `5` | Frame rate |
| **`cv2.CAP_PROP_FOURCC`** | `6` | 4-character code of codec |

Profundicemos nisto coa axuda dun exemplo.

#### <font style="color:rgb(200,0,0)">Nota </font>

Pode pasar que o cambio da propiedade dea resultados inesperados. Isto é porque a asignación de propiedades pode estar desabilitada para o ficheiro ou webcam. Por exemplo, podrías non ser capaz de asignar propiedades ao video dun ficheiro pero so a unha webcam. 

Incluso para a cámara web, no é posible establecer valores arbitrarios. Cualquiera valor que se estableza debe ser compatible co dispositivo. (Por exemplo, a resolución predeterminada da miña cámara web é 720x1280. Se quero configurarlo en 200x200, é posible que non suceda e se configure noutra cosa, poñamos por caso en 640x480, que é compatible coa cámara web.

````python
# anchura
ret = cap.set(3, 320)
print(ret)
# altura
ret = cap.set(4, 180)
print(ret)
````
