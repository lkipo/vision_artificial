### Funcin para visualizar imaxes en Jupyter Notebooks/Google Colab

````python
def plt_imshow(title, image):
        # convertemos a imaxe dende BGR ao espazo de cor RGB e a visualizamos
    if len(image.shape) == 3:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    #plt.imshow(image, cmap="gray")
    plt.imshow(image, cmap="jet")
    plt.title(title)
    plt.grid(False)
    plt.show()
````

### IImplementacin das funcins base de debuxo con OpenCV

````python
# Inicializamos o canvas cun mensura de 300x300 pixels e con 3 canles
# (R,G,B) con fondo negro
canvas = np.zeros((300, 300, 3), dtype="uint8")
````

# 2. Debuxando lias

<hr style = "border: none; height: 4px; background-color: # D3D3D3" />

A funcin **`line ()`** en OpenCV sase para debuxar lias nas imaxes.

### <font color = "green"> Sintaxe da funcina </font>
"Python
    img = cv2.line (img, pt1, pt2, cor [, espesor [, LineType [, shift]]])
""

**`img`** : a imaxe de sada que foi anotada.

A funcin ten ** 4 argumentos necesarios **:

1. `img`: Imaxe sobre a que trazaremos unha lia
2. `pt1`: primeiro punto (x, y) localizacin do segmento de lia
3. `pt2`: segundo punto do segmento de lia
4. `color`: cor da lia que se trazar
    
Os argumentos opcionais que se usan habitualmente inclen:

1. `espesor`: nmero enteiro que especifica o grosor da lia. O valor predeterminado  1.
2. `lineType`: tipo de lia. O valor predeterminado  ** `cv2.LINE_8` **, que significa unha lia de 8 conexins. Normalmente, **`cv2.LINE_AA`** (lia suave) sase para o lineType.
    
### <font color = "green"> Documentacin OpenCV </font>

[**`line()`**](https://docs.opencv.org/4.5.2/d6/d6e/group__imgproc__draw.html#ga7078a9fae8c7e7d13d24dac2520ae4a2)
[**`linetype`**](https://docs.opencv.org/4.5.2/d6/d6e/group__imgproc__draw.html#gaf076ef45de481ac96e0ab3dc2c29a777)

<hr style="border:none; height: 4px; background-color:#D3D3D3" />

````python
# debuxamos unha lia verde dende a esquina superior esquerda do noso canvas
# ata a esquina inferior dereita
green = (0, 255, 0)
cv2.line(canvas, (0, 0), (300, 300), green)
plt_imshow("Canvas", canvas)

# debuxamos unha lia de 3 pixels de gordo dende a esquena superior dereita ata a inferior esquerda
red = (0, 0, 255)
cv2.line(canvas, (300, 0), (0, 300), red, 3)
plt_imshow("Canvas", canvas)
````

# Debuxar un rectngulo

<hr style = "border: none; height: 4px; background-color: # D3D3D3" />

A funcin **`rectangle ()`** en OpenCV sase para debuxar rectngulos en imaxes.

### <font color = "green"> Sintaxe de funcins </font>

    img = cv2.rectngulo (img, pt1, pt2, cor [, espesor [, tipo de lia [, desprazamento]]])

`img`: a imaxe de sada que foi anotada.

A funcin ten **4 argumentos necesarios**:

1. `img`: imaxe sobre a que se debuxa o rectngulo.
2. `pt1`: vrtice do rectngulo. Normalmente usamos o **vrtice superior esquerdo** aqu.
3. `pt2`: vrtice do rectngulo oposto a pt1. Normalmente aqu usamos o vrtice **inferior dereito**.
4. `color`: cor do rectngulo
    
Os argumentos opcionais que se usan habitualmente inclen:

1. "espesor": espesor do contorno do crculo (se  positivo).
    Se se proporciona un valor negativo para este argumento, resultar un rectngulo cheo.
2. `lineType`: tipo do lmite do crculo. Isto  exactamente igual ao argumento lineType en
    **`cv2.line ()`**
    
### <font color = "green"> Ligazns de documentacin OpenCV </font>

[**`rectangle()`**](https://docs.opencv.org/4.5.2/d6/d6e/group__imgproc__draw.html#ga07d2f74cadcf8e305e810ce8eed13bc9)

<hr style="border:none; height: 4px; background-color:#D3D3D3" />

````python
# debuxamos un cadrado verde de 50x50 pixel, empezando en 10x10 e rematando en 60x60
cv2.rectangle(canvas, (10, 10), (60, 60), green)
plt_imshow("Canvas", canvas)

# debuxamos outra rectngulo, este vermello con 5 pixels de gordo
cv2.rectangle(canvas, (50, 200), (200, 225), red, 5)
plt_imshow("Canvas", canvas)

# debuxamos un rectngulo (azul e recheo!)
blue = (255, 0, 0)
cv2.rectangle(canvas, (200, 50), (225, 125), blue, -1)
plt_imshow("Canvas", canvas)
````

# Debuxar un crculo

<hr style = "border: none; height: 4px; background-color: # D3D3D3" />

A funcin **`circle ()`** en OpenCV sase para debuxar crculos en imaxes.

### <font color = "green"> Sintaxe de funcins </font>

```Python
img = cv2.circle (img, centro, radio, cor [, espesor [, tipo de lia [, desprazamento]]])
```

`img`: a imaxe de sada que foi anotada.

A funcin ten **4 argumentos necesarios**:

1. `img`: Imaxe sobre a que trazaremos unha lia
2. `center`: centro do crculo
3. `radio`: radio do crculo
4. `cor`: cor do crculo que se debuxar
    
Os argumentos opcionais que se usan habitualmente inclen:

1. `espesor`: espesor do contorno do crculo (se  positivo).
Se se proporciona un valor negativo para este argumento, resultar un crculo recheo.
2. `lineType`: tipo do lmite do crculo. Isto  exactamente igual ao argumento lineType en **`cv2.line ()`**
    
### <font color = "green"> Documentacin OpenCV </font>

[**`circle()`**](https://docs.opencv.org/4.5.2/d6/d6e/group__imgproc__draw.html#gaf10604b069374903dbd0f0488cb43670)

<hr style="border:none; height: 4px; background-color:#D3D3D3" />


````python
# reinicializamos o noso canvas como un array baleiro e achamos as coordenadas (x,y) do centro 
canvas = np.zeros((300, 300, 3), dtype="uint8")
(centerX, centerY) = (canvas.shape[1] // 2, canvas.shape[0] // 2)
white = (255, 255, 255)

# lazo no que incrementamos o radio, dende 25 pixels a 150 pixels incrementos de 25
for r in range(0, 175, 25):
	# debusamos un crculo branco co ltimo valor do radio
	cv2.circle(canvas, (centerX, centerY), r, white)

# amosamos a nosa obra de arte!
plt_imshow("Canvas", canvas)
````

````python
# reinicialozamos o noso canvas de novo
canvas = np.zeros((300, 300, 3), dtype="uint8")

# Debuxemos 25 crculos aleatorios
for i in range(0, 25):
	# xeramos aleatoriamente un radio entre 5 e 200 pixels, unha cor aleatoria,
	# e un punto, tamn aleatorio, para ser o centro do crculo que imos debuxar
	radius = np.random.randint(5, high=200)
	color = np.random.randint(0, high=256, size=(3,)).tolist()
	pt = np.random.randint(0, high=300, size=(2,))

	# debuxamos o crculo cos valores aleatorios
	cv2.circle(canvas, tuple(pt), radius, color, -1)

# visualizamos a nosa obra maestra
plt_imshow("Canvas", canvas)
````

# Engadindo texto

<hr style = "border: none; height: 4px; background-color: # D3D3D3" />

A funcin **`putText ()`** en OpenCV sase para engadir texto s imaxes.

### <font color = "green"> Sintaxe de funcins </font>
    img = cv2.putText (img, text, org, fontFace, fontScale, color [, espesor [, lineType [, bottomLeftOrigin]]])

`img`: a imaxe de sada que foi anotada.

A funcin ten **6 argumentos necesarios**:

1. `img`: imaxe na que hai que escribir o texto.
2. `texto`: cadea de texto a escribir.
3. `org`: esquina inferior esquerda da cadea de texto na imaxe.
4. `fontFace`: tipo de letra
5. `fontScale`: factor de escala de tipo de letra que se multiplica polo tamao base especfico da fonte.
6. `color`: cor do tipo de letra
 
Os argumentos opcionais que se usan habitualmente inclen:

1. `espesor`: nmero enteiro que especifica o grosor da lia para o texto. O valor predeterminado  1.
2. `lineType`: tipo de lia. O valor predeterminado  8, que significa unha lia conectada a 8. Normalmente, **`cv2.LINE_AA` ** (lia antialiased ou suave) sase para o lineType.

### <font color = "green"> Documentacin OpenCV </font>

[**`putText()`**](https://docs.opencv.org/4.5.2/d6/d6e/group__imgproc__draw.html#ga5126f47f883d730f633d74f07456c576)

<hr style="border:none; height: 4px; background-color:#D3D3D3" />

### Debuxando e escribindo texto sobre imaxes

````python
#empregamos un diccionario en vez de argumentos no Jupyter
# para definir os parmetros de entrada ao noso programa
args = {
	"image": "../data/images/beach.png"
}
````

````python
# cargamos a imaxe
image = cv2.imread(args["image"])

# debuxamos un circulo, dous circulos recheos e un rectnculo a modo emoticona
cv2.circle(image, (168, 188), 90, (0, 0, 255), 2)
cv2.circle(image, (150, 164), 10, (0, 0, 255), -1)
cv2.circle(image, (192, 174), 10, (0, 0, 255), -1)
cv2.rectangle(image, (134, 200), (186, 218), (0, 0, 255), -1)

#escribimos un texto na imaxe
cv2.putText(
     image, #imaxe onde escribimos (array numpy)
     "Ola onda!", #texto
     (30,30), #posicion na que empeza a escritura
     cv2.FONT_HERSHEY_SIMPLEX, #famila da fonte
     1, #tamao da fonte en puntos
     (0, 0, 255), #cor da fonte
     2) #anchura da fonte


# visualizamos
plt_imshow("Output", image)
````
