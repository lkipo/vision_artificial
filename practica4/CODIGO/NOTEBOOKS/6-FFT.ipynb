## A base de Fourier

Calquera sinal pdese expresar como unha combinacin de ondas de diferentes frecuencias.

$$ f(x) = \sum_\omega A_\omega \cos(2\pi \omega x + \phi_\omega) $$

Cada frecuencia $\omega$ achega unha certa amplitude $A_\omega$ e un desprazamento $\phi_\omega$. A Transformada de Fourier  un algoritmo que analiza o sinal e obtn estes dous ingredientes.

Matematicamente  moito mellor usar ondas **complexas**

$$ f(x) = \sum _ \omega F_\omega  e ^ {2 \pi i \omega x} $$

onde os coeficientes codifican nun nmero complexo a amplitude e o desprazamento de cada frecuencia.

$$ F_\omega = A e^{i \phi_\omega} $$

Hai unha explicacin mis detallada [neste documento] (https://robot.inf.um.es/material/va/percep.pdf#appendix.B), que tamn incle unha interpretacin xeomtrica da base de exponenciais complexas.

```python
# Importamos as librerias
import cv2
import scipy as sp
from scipy import signal
import numpy as np
import numpy.fft as fft
import matplotlib.pyplot as plt
%matplotlib inline
```

Como exemplo, xeramos un sinal con compoentes de frecuencia coecidos e vemos a FFT os atopa: 
    

```python
# compoente frecuencial de amplitude a, frecuencia w e fase d
def C(w,d,x):
    return np.cos(2*np.pi*w*x+d)

def fun(x):
    return 1 + 3*C(2,0.1,x) + 4*C(5,-0.2,x) + 2*C(15,0.3,x)

x = np.linspace(0,1,1000)

plt.figure(figsize=(12,4))
plt.plot(x,fun(x),lw=2);
plt.grid();
```

```python
#Conseguimos a sinal sunamdo estes catro armnicos
plt.figure(figsize=(12,4))
plt.plot(x,1*C(0,0,x), label='freq 0')
plt.plot(x,3*C(2,0.1,x), label='freq 2')
plt.plot(x,4*C(5,-0.2,x), label='freq 5')
plt.plot(x,2*C(15,0.3,x), label='freq 15');
plt.grid(); plt.legend();
```

Na prctica, s observamos un mostreo da sinal

```python
#xeramos a sinal mostreada cun intervalo delta
delta = 0.02
xs =  np.linspace(0,1-delta,50)
f = fun(xs)

print(f)

plt.figure(figsize=(12,4))
plt.plot(x,fun(x),'r',xs,f,'.b',markersize=10); plt.grid()
plt.title('sinal mostreada en 50 puntos');
```

O algoritmo FFT (transformada discreta de Fourier rpida) analiza o sinal mostreado e devolve as amplitudes e fases das compoentes, anda que o fai dun xeito un tanto especial, codificando o mdulo e o desfase como nmeros complexos situados en das posicins da matriz resultante.

**RECORDATORIO:** Numpy admite nmeros complexos.

 Dado un nmero complexo `x`:

     np.complex() constre un nmero complexo
     x.real refrese a compoente real
     x.imag refrese a compoente imaxinario
     np.abs(x) calcula a magnitude (ou amplitude)
     np.angle(x) calcula o ngulo en radianes (ou fase) 

```python
# dividimos por n para que a escala global da FFT coincida coas amplitudes
F = fft.fft(f)/len(f)

# para visualizar mellor o array, quitamos os ceros numricos
F[abs(F)<1E-10] = 0

F
```

Imos manipular un pouco a matriz para extraer os compoentes do sinal. O mdulo do resultado:

```python
abs(F)
```

```python
w = np.array(range(len(F)))

plt.figure(figsize=(18,4))
plt.bar(w,abs(F), color='green',alpha=0.5)
plt.xticks(w); plt.axis([-1,50,0,2.1]);
plt.title('O mdulo da FFT codifica a amplitude');
```

E a fase:

```python
w = np.array(range(len(F)))

plt.figure(figsize=(18,4))
plt.bar(w,np.angle(F), color='brown',alpha=0.5)
plt.xticks(w); plt.axis([-1,50,-0.5,0.5]);
#plt.grid()
plt.title('A fase da FFT codifica o desprazamento');
plt.grid(axis='y')
```

Vemos que a magnitude da transformada de Fourier, que  un sinal complexo, contn a amplitude de cada onda $ A_\omega $ e a "fase"  o desprazamento $ \phi_\omega $. Ten en conta que cada frecuencia $ \omega $ aparece na transformacin en das posicins, correspondentes a $ \omega $ e $ - \omega $ (que en realidade son a mesma onda) e que as frecuencias negativas aparecen ao final do vector.

O FFT  peridico e, polo tanto, as frecuencias superiores a $ N / 2 $ (o mximo que se pode detectar) son equivalentes a frecuencias negativas. O normal  amosar o fft "volteado" ou "centrado" para poer as baixas frecuencias no centro do vector.

```python
n = len(F)
n2 = (n+1) // 2
w = np.array(range(len(F)))
ws = np.roll(w,n2)
Fs = np.roll(F,n2)


plt.figure(figsize=(18,4))
plt.bar(w-25,abs(Fs), color='green',alpha=0.5)
plt.xticks(w-25); 
plt.axis([-24,24,0,2.5]);
plt.title('abs(FFT)');
```

Cando o sinal  real, como neste exemplo, a FFT ten simetra complexa conxugada. As frecuencias negativas teen a mesma amplitude absoluta cas positivas, polo que podemos visualizar soamente as frecuencias positivas: 

```python
n2 = len(F)//2
P = abs(F)[:n2]
P[2:] *= 2
plt.figure(figsize=(10,4))
plt.bar(w[:n2],P, color='blue',alpha=0.5)
plt.xticks(w[:n2]); 
plt.axis([-.5,24.5,0,5]);
plt.title('espectro de frecuencias');
```

Estas amplitudes coinciden con los coeficientes utilizados para definir `fun`. 
Se se toman $ n $ mostras, s se pode analizar ata a frecuencia $ n \over 2 $. 


# <font style="color:rgb(50,120,229)">Teorema de Parseval ou potencia espectral</font>
A enerxa da sinal reprtese entre as diferentes frecuencias:
$$\int_t \left|x(t)\right|^2 dt = \int_\omega \left|F(\omega)\right|^2 d \omega$$

### Transformada inversa

Podemos sintetizar ondas creando un vector coas amplitudes de cada frecuencia e aplicando a transformada inversa:

```python
F = np.zeros(100,dtype=complex)
F[3] = 2
F[5] = 1
f = fft.ifft(F)*len(F)
plt.plot(np.real(f));
```

### Teorema de Nyquist (ou teorema de mostraxe) 

O teorema de Nyquist afirma que, para reproducir adecuadamente un sinal, debe ser mostrado periodicamente a unha velocidade que  das veces a frecuencia mis alta que se desexa representar. 


```python
t = np.linspace(0, 1, 1000)

a = 1
f = 10 # frecuencia da sinal
p = np.pi/2
omega = 2 * np.pi * f
y = a * np.sin( omega * t + p)

plt.figure(figsize=(10, 4))
plt.title(f'Seno para a amplitude={a}, f={f}, e $\phi$={p}')
plt.plot(t, y, 'k')
plt.xlabel('$t$')
plt.ylabel('$a \sin(2 \pi \omega t + \phi)$');
```

## Mostreo (conversin analxico-dixital)

A mostreo realzase para realizar conversin analxica a dixital. 

```python
Fs = 14 # frecuencia de mostreo
T = 1/Fs # periodo de mostreo
sampling_duration = 1 # segundos de mostreo
N = Fs * sampling_duration + 1 # numero total de puntos na sinal

t_sampled = np.arange(N)*T
y_discretized = a * np.sin( omega * t_sampled + p)

#Visualizamos
plt.figure(figsize=(10, 4))
plt.title(f'Valores mostreados supepostos a onda senoida con amplitude={a}, $\omega$={f}, and $\phi$={p}')
plt.plot(t, y, 'k')
plt.scatter(t_sampled, y_discretized, color='red')
plt.xlabel('$t$')
plt.ylabel('$a \sin(2 \pi \omega t + \phi)$');
```

## Reconstrucin (conversin dixital a analxica)

Usamos unha das moitas tcnicas de interpolacin dispoibles para reconstrur o sinal analxico.

```python
from scipy import interpolate

method = 'previous'
method = 'cubic'
y_reconstructed_from_samples1 = interpolate.interp1d(t_sampled, y_discretized, kind=method)(t)

#Visualizamos a reconstrucion da sinal a partir do mostreo
plt.figure(figsize=(10, 4))
plt.title(f'Seno con amplitude={a}, $\omega$={f}, and $\phi$={p:3.2} (mtodo de interpolacin="{method}")')
plt.plot(t, y, 'k', label='Sinal Orixinal', alpha=0.3)
plt.scatter(t_sampled, y_discretized, color='red', label='mostras')
plt.plot(t, y_reconstructed_from_samples1, 'b', label='reconstrucin a partir das mostras')
plt.xlabel('$t$')
plt.legend()
plt.ylabel('$a \sin(2 \pi \omega t + \phi)$');
```

## Teorema da convolucin

A transformada de Fourier da convolucin de das funcins  o produto das sas transformadas de Fourier.
$$\mathcal{F}\{g \ast h\} = \mathcal{F}\{g\} \ast \mathcal{F}\{h\}$$

A transformada de Fourier inversa do produto de das transformadas de Fourier  a convolucin das das transformadas de Fourier inversas.
$$\mathcal{F}^{-1}\{g h\} = \mathcal{F}^{-1}\{g\} \ast \mathcal{F}^{-1}\{h\}$$

A convolucin no dominio espacial equivale  multiplicacin no dominio da frecuencia.

## Propiedades da transformada de Fourier

- A transformada de Fourier dunha sinal real  simtrica ao redor da orixe.
- A enerxa da sinal  a mesma que a enerxa da sa transformada de Fourier.
- Linealidade 

$$\mathcal{F}\{a g(x) + b h(x)\} = a \mathcal{F}\{g(x)\} + b \mathcal{F}\{h(x)\}$$

# Convolucin de imaxes usando a transformada de Fourier

## Paso 1: constre un ncleo gaussiano 2D 

```python
import scipy.stats as st

def gkern(kernlen=5, sig=3):
    """Devolve un kernel Gaussiano 2D."""
    x = np.linspace(-kernlen/(3*sig), kernlen/(3*sig), kernlen+1)
    kern1d = np.diff(st.norm.cdf(x))
    kern2d = np.outer(kern1d, kern1d)
    return kern2d/kern2d.sum()



g = gkern(kernlen=11, sig=1)
plt.imshow(g)

```

## Paso 2: Facemos que este kernel tea o mesmo tamao que a imaxe reenchendo ceros 

```python
h, w = g.shape
g_resized = np.zeros(img.shape)
g_resized[:h,:w] = g

plt.figure(figsize=(8,8))
plt.imshow(g_resized)
```

## Paso 3: Toma a sa transformada de Fourier 

```python
g_fourier = np.fft.fft2(g_resized)
g_fourier_shifted = np.fft.fftshift(g_fourier)
g_spectrum_magnitude = 20 * np.log(np.abs(g_fourier_shifted))

#Visualizamos
plt.figure(figsize=(8,8))
plt.imshow(g_spectrum_magnitude, cmap = 'gray')
plt.title('Magnitude da FFT')
plt.xticks([])
plt.yticks([]);
```

## Paso 4: multiplica as das transformadas de Fourier 

```python
multiplied_fourier = img_fourier * g_fourier
```


## Paso 5: Toma a transformada de Fourier inversa. 

```python
img_filtered = np.real(np.fft.ifft2(multiplied_fourier))


#Visualizamos
plt.figure(figsize=(8,8))
plt.title('Convolucin empregando a TF')
plt.imshow(img_filtered, cmap='gray');

```

## Paso 6: Imos confirmar que obtemos o mesmo resultado se usamos a convolucin en dominio espacial. 

```python
img_filtered2 =  sp.signal.convolve2d(img, g, mode='same', boundary='fill')


#Visualizamos
plt.figure(figsize=(8,8))
plt.title('Convolucin no dominio espacial')
plt.imshow(img_filtered2, cmap='gray');

```

## Aplicacin: orientacin dominante dunha imaxe (moi simple!)

Nesta aplicacin, queremos detectar cal  a orientacin dominante nunha imaxe moi simple (con imaxes reais, o algoritmo pode dar resultados non desexados).


```python
#Xeramos unha imaxe sintetica e xiramola un ngulo coecido (o resultados do algoritmo)

hW, hH = 600, 300
hFreq = 10.5

# sobre unha imaxe de  [0,1)x[0,1)
x = np.linspace( 0, 2*hW/(2*hW +1), 2*hW+1)     # cols
y = np.linspace( 0, 2*hH/(2*hH +1), 2*hH+1)     # filas

[X,Y] = np.meshgrid(x,y)
A = np.sin(hFreq*2*np.pi*X)

plt.imshow(A, cmap = 'gray');
H,W = np.shape(A)
```

```python
from PIL import Image
A4 = 0.5*(1+ A)                           # Normalizamos entre  [0,1]
I = Image.fromarray(np.uint8(A4*255))     # Ocnvertimos a unha imaxe PIL

rot_angle = -65
B = I.rotate(rot_angle, expand = True)   # rotoamos sen recortar
img=np.asarray(B,dtype=np.float32)

#Dimension e centro da imaxe
rows, cols = img.shape
crow, ccol = rows//2, cols//2

#Achamos a TF
img_fourier = np.fft.fft2(img)/(rows*cols)
img_fourier_shifted = np.fft.fftshift(img_fourier)
img_spectrum_magnitude = np.abs(img_fourier_shifted)

#Visualizamos
plt.figure(figsize=(8,8))
#plt.imshow(img_spectrum_magnitude, cmap = 'gray')
plt.imshow(img_spectrum_magnitude[crow-25:crow+25,ccol-25:ccol+25], extent=[-25,25,-25,25]);
plt.title('Espectro da magnitude')
plt.xticks([])
plt.yticks([]);


#Debido a que p alto valor do espectro de potencia na orixe  alto, 
#enmascara os valores das frecuencias importantes. Eliminmolo restando o valor medio da imaxe. 
img_fourier = np.fft.fft2(img-img.mean())/(rows*cols)
img_fourier_shifted = np.fft.fftshift(img_fourier)
img_spectrum_magnitude = np.abs(img_fourier_shifted)


#Visualizamos
plt.figure(figsize=(8,8))
#plt.imshow(img_spectrum_magnitude, cmap = 'gray')
plt.imshow(img_spectrum_magnitude[crow-25:crow+25,ccol-25:ccol+25], extent=[-25,25,-25,25]);

plt.title('Espectro da magnitude')
plt.xticks([])
plt.yticks([]);
```

```python
#Calculamos as coordenadas da compoente de frecuencia mis grande e atopamos o ngulo 

indices = np.where(img_spectrum_magnitude == np.max(img_spectrum_magnitude))
maxY = (indices[0][0]-crow)/rows 
maxX = (indices[1][0]-ccol)/cols

alpha=np.arctan(maxY/maxX)*180/np.pi

print(alpha)
```

```python
# Despois, xiramos a imaxe orixinal correspondentemente 

C =B.rotate(alpha)
plt.figure()
plt.imshow(C,cmap = 'gray');

```
