# 1. Introducin aos histogramas

Os histogramas recollen os recontos de datos organizados nun conxunto de contedores predefinidos. Cando achamos histogramas, precisamos especificar o nmero de contedores ao longo do eixo x que representa un intervalo de valores de intensidades de pxeles: 0-9, 10-19, 20-29, etc. A continuacin veremos varios exemplos que axudarn a asentar este concepto.

<hr style="border:none; height: 4px; background-color:#D3D3D3" />

### <font color="green">Sintaxe da funcin </font>
``` python
retval = plt.hist(x[, bins[, range[, ...]]])
```

`retval`: Matriz de datos para realizar o histograma. Tamen pode ser secuencia de matrices [datos1, datos2, ...]. O tipo da matriz sempre ser flotante anda que non se utilice ningunha ponderacin ou normalizacin.

A funcin ten **1 argumento de entrada obrigatorio** e varios indicadores opcionais:

1. `x`: matriz ou secuencia de matrices para realizar o histograma.
2. `bins`: define o nmero de intervalos nos que se mostrea o rango de datos. Este  un **argumento opcional** cun valor predeterminado de 10.
2. `range`: o rango inferior e superior dos intervalos. Este  un **argumento opcional** cun valor predeterminado de None, equivalente a usar todo o rango dos datos de entrada `x`.

### <font style="color:rgb(50,120,230)">Creamos un histograma dunha imaxe negra</font>

````python
# Creamos un array de dimensions 20x20x1 rechea de zeros.
black_img = np.zeros([20, 20, 1])

# Aplanamos os datos da imaxe nun array 1D.
black_flatten = black_img.ravel()

# Visualizamos a imaxe e o histograma.
plt.figure(figsize=[18, 4])

plt.subplot(131); plt.imshow(black_img); plt.title('Imaxe negra')

plt.subplot(132) 
plt.hist(black_flatten, range = [0, 256])
plt.xlim([0, 256])
plt.xlabel('Intensidade dos pixels')
plt.ylabel('Numero de pixels')
plt.title('Histograma dunha imaxe negra')
````

### <font style="color:rgb(50,120,230)">Explicacin </font>

Usando a funcin de ceros numerosos, podemos xerar facilmente unha simple imaxe negra. Tcnicamente  unha imaxe en escala de grises cunha soa canle, inclundo un total de 400 pxeles. A continuacin, creamos un histograma desta imaxe usando o nmero predeterminado de contedores (10) e especificamos que o histograma debera crear estes contedores de xeito uniforme a travs de valores de intensidade de 0 a 256. O extremo superior deste intervalo non se incle, polo que o o histograma pode mostrar en ltima instancia calquera intensidade de pxel ata un valor de 255 includo.

As barras verticais do histograma representan o nmero total de pxeles cun valor de intensidade que se atopa dentro dese colector en particular. Vendo o histograma resultante dunha imaxe puramente negra, atopamos que todos os valores mapeados no primeiro colector. Tamn podemos ver que o nmero total de pxeles (400) equivale exactamente ao reconto de pxeles do valor y nesta primeira papeleira, relacionado co tamao da imaxe 20x20 que xeramos.

### <font style="color:rgb(50,120,230)">Histrograma dunha imaxe binaria</font>

````python
# Lemos a imae.
img = cv2.imread('./data/checkerboard_18x18.png', cv2.IMREAD_GRAYSCALE)

# Aplanamos a imaxe nun array 1D.
img_flatten = img.ravel()

# Visualizamos as imaxes e os histogramas
plt.figure(figsize = [18, 4])

plt.subplot(131); plt.imshow(img); plt.title('Imaxe orixnal')

plt.subplot(132) 
plt.hist(img_flatten, 5, [0, 256])
plt.xlim([0, 256])
plt.xlabel('Intensidade dos pixels')
plt.ylabel('Numero de pixels')
plt.title('5 contedores (Bins)')

plt.subplot(133)
plt.hist(img_flatten, 50, [0, 256])
plt.xlim([0, 256])
plt.xlabel('Intensidade dos pixels')
plt.ylabel('Numero de pixels')
plt.title('5o contedores (Bins)')
````

### <font style = "color: rgb (50,120,230)"> Explicacin </font>

A imaxe de xadrez  de 18x18 pxeles, polo que o nmero total de pxeles na imaxe  de 324. Ademais, as intensidades dos pxeles son s 0 (negro) ou 255 (branco). Os grficos do histograma  dereita mostran a distribucin das intensidades dos pxeles. Cando trazamos histogramas,  til especificar o nmero de contedores (bins) ao longo do eixo x. O primeiro histograma ten 5 e o segundo histograma ten 50. Dado que estamos trazando a distribucin das intensidades de pxeles, o eixo x vara de 0 a 255. As, o ancho dos internvalos na primeira grfica  51 e na segunda grfica  5,1 (non enteiro).

Neste exemplo, a altura da barra  esquerda de cada trazado  180. Isto significa que hai 180 pxeles con intensidades no rango indicado. Dado que as intensidades desta imaxe son 0 ou 255, o ancho do contedor (5.1 ou 51) non importa porque mbolos rangos inclen o valor de 0. Do mesmo xeito, o nmero total de pxeles brancos (144) coincide coa altura da barra na grfica da dereita (144), posto que os ltimos contedores conteen o valor de 255.

Os histogramas revelan que a relacin entre os cadrados branco e negro  de 5:4 (ou 180:144). Paga a pena resaltar que os histogramas deste tipo non conteen informacin espacial sobre a imaxe.

Seleccionar un tamao de contedor particular ten que ver coa granularidade, pero para as imaxes  moi comn establecer o nmero de contedores en 256, o que equivale a un ancho de contedor de 1.

### <font style="color:rgb(50,120,230)">Compara calcHist() con plt.hist()</font>


<hr style="border:none; height: 4px; background-color:#D3D3D3" />

### <font color="green">Sintaxe da Funcin </font>

```python
    hist = cv2.calcHist(images, channels, mask, histSize, ranges[, hist[, accumulate]])
```

Esta funcin ten **5 argumentos necesarios**:

1. `images`: Imaxes de entrada. Todas deben ter a mesma profundidade, CV_8U, CV_16U ou CV_32F, e o mesmo tamao. Cada unha delas pode ter un nmero arbitrario de canles.

2. `channels`: lista das canles usadas para calcular o histograma.

3. `mask`: mscara opcional. Se a matriz non est baleira, debe ser unha matriz de 8 bits do mesmo tamao que as images[0]. Os elementos de mscara distintos de cero marcan os elementos da matriz contados no histograma.

4. `histSize`: matriz de tamaos de histogramas en cada dimensin.


### <font color="green">Documentacin de OpenCV</font>


[**`calcHist()`**](https://docs.opencv.org/4.5.2/d6/dc7/group__imgproc__hist.html#ga4b2b5fd75503ff9e6844cc4dcdaed35d)

<hr style="border:none; height: 4px; background-color:#D3D3D3" />

````python
# Lemos a imaxe
img = cv2.imread('./data/Apollo-8-Launch.jpg', 0)

# Usamos calcHist() en OpenCV.
hist = cv2.calcHist(images = [img], channels = [0], mask = None, histSize = [256], ranges = [0,255])

# Aplanado da imaxe
img_flatten = img.ravel()

# Visualizacipn
plt.figure(figsize = [18, 4])
plt.subplot(131); plt.imshow(img); plt.title('Imaxe Orixinal')
plt.subplot(132); plt.plot(hist); plt.xlim([0, 256]); plt.title('cv2.calcHist()')
plt.subplot(133); plt.hist(img_flatten,256,[0,256]); plt.xlim([0, 256]); plt.title('np.ravel(), plt.hist()')
````

### <font style="color:rgb(50,120,230)">Diferentes imaxes con mesmo histograma</font>

# 2. Histogramas de cor

````python
# Lemos as imaxes de cor
img = cv2.imread('./data/Emerald_Lakes_New_Zealand.jpg')

# Histragramas para cada banda de cor
hist1 = cv2.calcHist([img], [0], None, [256], [0, 255])
hist2 = cv2.calcHist([img], [1], None, [256], [0, 255])
hist3 = cv2.calcHist([img], [2], None, [256], [0, 255])

# Visualizacion
plt.figure(figsize = [18, 10])
plt.subplot(221); plt.imshow(img[:, :, ::-1])

plt.subplot(222) 
plt.plot(hist1, 'b'); plt.plot(hist2, 'g'); plt.plot(hist3, 'r') 
plt.xlim([0, 256])
plt.ylim([0, 200000])
````

### <font style="color:rgb(50,120,230)">Empregando unha mascara con calcHist()</font>

````python
# Lemos as imaxes.
img = cv2.imread('./data/Emerald_Lakes_New_Zealand.jpg')

# Creamos unha mascara para filtrar unha imaxe para calculo de histograma.
mask_hist = np.zeros((img.shape[0], img.shape[1]), dtype = 'uint8')

# Selecionamos a rexion para illar o lago
mask_hist[2100:2400, 1500:2200] =  255

# Crea unha mscara similar para mostrar a rexin seleccionada na imaxe (s para fins de visualizacin).
# A mscara debe ter o mesmo nmero de canles de cor que a imaxe, pero cada canle de cor ter
# conter a mesma informacin. 
mat = [mask_hist, mask_hist, mask_hist]
mask_3ch = cv2.merge(mat, 3)

# Crea unha imaxe que s amose a rexin de interese seleccionada. 
img_roi = cv2.bitwise_and(img, mask_3ch)

# Calculamos histogramas para cada canle de cor para ambas as imaxes. 
hist1_lake = cv2.calcHist([img], [0], mask_hist, [256], [0, 255])
hist2_lake = cv2.calcHist([img], [1], mask_hist, [256], [0, 255])
hist3_lake = cv2.calcHist([img], [2], mask_hist, [256], [0, 255])

# Visualizacion
plt.figure(figsize = [18, 10])
plt.subplot(223); plt.imshow(img_roi[:, :, ::-1])
plt.subplot(224); plt.plot(hist1_lake, 'b'); plt.plot(hist2_lake, 'g'); plt.plot(hist3_lake, 'r') 
plt.xlim([0, 256])
plt.ylim([0, 10000])
````

# 3. Igualacin do histograma
Un histograma de imaxe  unha representacin grfica da distribucin tonal dos datos. Cada histograma  simplemente unha matriz con 256 bins e cada bin contn o nmero de pxeles con esa intensidade.
A igualacin do histograma  un mtodo non lineal para mellorar o contraste nunha imaxe. Vexamos como realizar a ecualizacin do histograma en OpenCV usando [**`equalizeHist()`**](https://docs.opencv.org/4.1.0/d6/dc7/group__imgproc__hist.html#ga7e54091f0c937d49bf84152a16f76d6e). 

## 3.1 Ecualizacin de hostogramas en imaxes de gris</font>

A funcin `equalizeHist()` realiza a igualacin do histograma nunha imaxe en escala de grises. A sintaxe dse a continuacin. 

<hr style="border:none; height: 4px; background-color:#D3D3D3" />

### <font color="green">Sintaxe da Funcin </font>

```python
	dst = cv2.equalizeHist(src[, dst])
```

**Parameters**

- **`src`**:	Imaxe de canle de 8 bits de orixe. 
- **`dst`**:	Imaxe de destino do mesmo tamao e tipo que src. 

### <font color="green">Documentacin OpenCV </font>


[**`equalizeHist()`**](https://docs.opencv.org/4.1.0/d6/dc7/group__imgproc__hist.html#ga7e54091f0c937d49bf84152a16f76d6e)

<hr style="border:none; height: 4px; background-color:#D3D3D3" />

````python
# Lemos a imaxe de gris
img = cv2.imread('./data/parrot.jpg', cv2.IMREAD_GRAYSCALE)

# Visualizamos
plt.figure(figsize = (18, 6))
plt.subplot(121); plt.imshow(img); plt.title('Imaxe Orixinal')
````

````python
# Lemos a imaxe
img = cv2.imread('./data/parrot.jpg', cv2.IMREAD_GRAYSCALE)

# Ecualizamos o histograma
img_eq = cv2.equalizeHist(img)

# Visualizacion
plt.figure(figsize = (18, 6))
plt.subplot(121); plt.imshow(img_eq); plt.title('Imaxe ecualizada')
plt.subplot(122); plt.hist(img_eq.ravel(), 256, [0, 256]); plt.title('Histograma ecualizada')
````

Como era de esperar, o histograma reprtese de xeito mis uniforme no rango 

## 3.2 Igualacin do histograma para imaxes en cor </font>

Para imaxes en cor, non podemos aplicar a igualacin do histograma nas canles R, G, B por separado. Para entender por que non  moi boa idea, botemos un ollo a un exemplo.

### <font style="color:rgb(50,120,230)">Forma incorrecta</font>

````python
# Lemos a imaxe decor
img = cv2.imread('./data/parrot.jpg')
img_eq = np.zeros_like(img)

# Ecualizacion de cada canle por separado
for i in range(0, 3):
    img_eq[:, :, i] = cv2.equalizeHist(img[:, :, i])

# Visualizacion
plt.figure(figsize = (18, 6))
plt.subplot(121); plt.imshow(img[:, :, ::-1]); plt.title('Imaxe orixianl de cor')
plt.subplot(122); plt.imshow(img_eq[:, :, ::-1]); plt.title('Imaxe ecualizada')
````

### <font style="color:rgb(50,120,230)">Forma correcta</font>

Acabamos de ver que a igualacin do histograma realizada nas tres canles por separado leva a resultados pobres. A razn  que cando cada canle de cor se transforma de forma non lineal independentemente, pode obter cores completamente novas e non relacionadas.

O xeito correcto de realizar a igualacin do histograma en imaxes en cor  transformar as imaxes nun espazo como o espazo de cores ** HSV ** onde as cores estn separadas da intensidade.

** FLUXO DE TRABALLO **

1. Transforma a imaxe ao espazo de cores HSV.
2. Realice a ecualizacin do histograma s na canle V.
3. Transforma a imaxe de novo ao espazo de cores RGB.

````python
# Lemos a imaxe
img = cv2.imread('./data/parrot.jpg', cv2.IMREAD_COLOR)

# Convertemos a HSV
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Ecualizamos somanete a canle V para o valor de intensidade.
img_hsv[:,:,2] = cv2.equalizeHist(img_hsv[:, :, 2])

# Retornamos ao formato BGR.
img_eq = cv2.cvtColor(img_hsv, cv2.COLOR_HSV2BGR)

# Visualizamos
plt.figure(figsize = (18, 6))
plt.subplot(121); plt.imshow(img[:, :, ::-1]); plt.title('Imaxe orixinal de cor')
plt.subplot(122); plt.imshow(img_eq[:, :, ::-1]); plt.title('Imaxe ecualizada')
````

````python
#Visualizamos os histogramas
plt.figure(figsize = [15,4])
plt.subplot(121); plt.hist(img.ravel(),256,[0,256]); plt.title('Imaxe orixinal')
plt.subplot(122); plt.hist(img_eq.ravel(),256,[0,256]); plt.title('Histograma ecualizado')
````

## <font style = "color: rgb (50,120,229)"> Contraste limitado de adaptacin do histograma de igualacin (CLAHE) </font>

A igualacin do histograma usa os pxeles de toda a imaxe para mellorar o contraste. Anda que isto poida parecer bo en moitos casos, s veces  posible que queiramos mellorar o contraste localmente para que a imaxe non pareza mis natural e menos dramtica.

Para tales aplicacins, usamos [Contrast Limited Adaptive Histogram Equalization (CLAHE)] (https://docs.opencv.org/4.1.0/d5/daf/tutorial_py_histogram_equalization.html)) que mellora o contraste local. Mis importante anda, permtenos especificar o tamao do vecindario que se considera "local".

Escollamos unha imaxe diferente, onde podemos preferir CLAHE en lugar da igualacin regular do histograma. 

````python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

#Reconfiguramos as opcion de matplotlib
import matplotlib
matplotlib.rcParams['figure.figsize'] = (50.0, 50.0)
matplotlib.rcParams['image.cmap'] = 'gray'
matplotlib.rcParams['axes.titlesize'] = 40
matplotlib.rcParams['image.interpolation'] = 'bilinear'

# Lemos a imaxe de cor
filename = "./data/night-sky.jpg"
im = cv2.imread(filename)

# Convertemos a HSV 
imhsv = cv2.cvtColor(im, cv2.COLOR_BGR2HSV)
imhsvCLAHE = imhsv.copy()

# Ecualizamos so a canle V 
imhsv[:,:,2] = cv2.equalizeHist(imhsv[:,:,2])

clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
imhsvCLAHE[:,:,2] = clahe.apply(imhsvCLAHE[:,:,2])

# Retornamos ao espazo BGR 
imEq = cv2.cvtColor(imhsv, cv2.COLOR_HSV2BGR)
imEqCLAHE = cv2.cvtColor(imhsvCLAHE, cv2.COLOR_HSV2BGR)


#Visualizamos as imaxes
plt.figure(figsize=(40,40))

ax = plt.subplot(1,3,1)
plt.imshow(im[:,:,::-1], vmin=0, vmax=255)
ax.set_title("ImaxeOrixinal")
ax.axis('off')


ax = plt.subplot(1,3,2)
plt.imshow(imEq[:,:,::-1], vmin=0, vmax=255)
ax.set_title("Histograma ecualizado")
ax.axis('off')

ax = plt.subplot(1,3,3)
plt.imshow(imEqCLAHE[:,:,::-1], vmin=0, vmax=255)
ax.set_title("CLAHE")
ax.axis('off')
````
