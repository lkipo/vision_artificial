## <font style="color:rgb(50,120,229)">Histogramas de cores</font>     
     
   
A estas alturos do curso, debes saber o que  un histograma, concretamente un histograma de cor. Pero repasarmolos brevemente por se acaso. En pocas palabras, un histograma de cor conta o nmero de veces que unha determinada intensidade de pxeles (ou rango de intensidades de pxeles) ocorre nunha imaxe. Usando un histograma de cor podemos expresar a distribucin real ou **cantidad** de cada cor nunha imaxe. Os recontos para cada intervalo de cor/cor sanse entn como o noso vector de caractersticas.

Como descritores de imaxes, os histogramas de cores tenden a ser bastante potentes. Se decidimos utilizar un histograma de cores **3D** con 8 caixas (bins) por canle, poderiamos representar calquera imaxe de calquera tamao usando s **8 x 8 x 8 = 512 caixas**, ou un vector de caractersticas de **512-D**. O tamao dunha imaxe non ten ningn efecto no noso histograma de cor de sada, anda que  conveniente cambiar o tamao das imaxes grandes a unha dimensin mis manexable para aumentar a velocidade do clculo do histograma.

## <font style="color:rgb(50,120,229)">Clusterin k-medias</font>
K-means  un algoritmo de agrupacin onde o seu obxectivo  particionar **n puntos** de datos en **k clusters**. Cada un dos n puntos de datos asignarase ao clster (coa media) mis prxima. A media de cada grupo chmase o seu "centroide" ou "centro":


Un exemplo de k-means que separan datos 2-d en 3 grupos distintos.

En xeral, a aplicacin de k-medias obtn k grupos separados dos n puntos de datos orixinais. Os puntos de datos dentro dun clster particular considranse "mis similares" entre si que os puntos de datos que pertencen a outros clsteres.

Nesta exemplo, agruparemos os histogramas de cores extrados das imaxes do noso conxunto de datos, pero en realidade, podes agrupar calquera tipo de vector de caractersticas.

Os histogramas que pertencen a un grupo dado sern mis similares na distribucin de cores que os histogramas pertencentes a un grupo separado.

Unha advertencia de k-means  que **necesitamos especificar o nmero de clusters** que queremos xerar antes inicialmente. Hai algoritmos que seleccionan automaticamente o valor ptimo de k, pero estes algoritmos non os empregaremos neste curso. Polo momento, proporcionaremos manualmente un valor de k=2 para separar as das clases de imaxes.

## <font style="color:rgb(50,120,229)">Vexamos estas ideas en accin ...</font>

Primeiro imos definir o noso descritor de imaxe e crearemos unha clase para el (poderiamos meter todos os descritores deste tema dentro dela)chamado LabHistogram para extraer histogramas de cores de imaxes no espazo de cor L*a*b*:

```python
# importamos paquetes
import cv2

class LabHistogram:
	def __init__(self, bins):
		# Almacenamos os bins que queremos para o histograma
		self.bins = bins

	def describe(self, image, mask=None):
		# Pasamos ao espazo de cor L*a*b*, achamos o histograma e normalizamolo
		lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
		hist = cv2.calcHist([lab], [0, 1, 2], mask, self.bins,
			[0, 256, 0, 256, 0, 256])

		# Histograma para versin de OpenCV 3+
		hist = cv2.normalize(hist,hist).flatten()

		# devolvemos o histograma
		return hist
```

A nosa clase, LabHistogram,  inicializada cun s un argumento necesario: o nmero de bins (caixas) no noso histograma 3D.
O mtodo describe() que acepta dous argumentos:  a imaxe da queremos extraer caractersticas e **unha mscara opcional, que nos permitira extraer caractersticas s das partes dunha imaxe que nos interesan**. Se empregas esta opcin, teras un descriptor local.

O primeiro paso  converter a nosa imaxe do espazo de cor RGB ao espazo de cor L*a*b*.  moi importante que realicemos o paso de normalizacin. Se non o fixeramos, entn as imaxes co mesmo contido pero diferentes tamaos teran histogramas dramticamente diferentes. Pola contra, ao normalizar o noso histograma asegurmonos de que o ancho e o alto da nosa imaxe de entrada **non tean ningn efecto** no histograma de sada. Finalmente, devolve o noso vector de caractersticas do histograma.

Por que estamos a usar o espazo de cor L*a*b*? a razn  a que distancia euclidiana entre das cores en L*a*b* ten un significado **perceptivo e perceptible**. E dado que o algoritmo de agrupacin de k-means asume un espazo euclidiano, obteremos mellores clusters usando o espazo de cor L*a*b* que RGB ou HSV.


```python
# Importacions estandar
from sklearn.cluster import KMeans
from imutils import paths  #pip install imutils
import numpy as np
import cv2
import numpy as np;
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12.0, 12.0)
matplotlib.rcParams['image.cmap'] = 'gray'
```

```python
# argumentos de onde estan os datos e o numero de clusters
args={'dataset':'./dataset/', 'clusters':2}

# Creamos o obxecto que describe a imaxe por un histograma
# de cor e fixamos o numero de bis desexado
desc = LabHistogram([8, 8, 8])
data = []

# almacenamos os path das imaxes do directorio dataset
imagePaths = list(paths.list_images(args["dataset"]))
imagePaths = np.array(sorted(imagePaths))

# lazo sobre todas as imaxes do dataser
for imagePath in imagePaths:
    # cargamos a imaxe, describimola e metemola na lista
    image = cv2.imread(imagePath)
    hist = desc.describe(image)
    data.append(hist)

# Facemos un clustering sobre os histogramas de cor
clt = KMeans(n_clusters=args["clusters"], random_state=42)
labels = clt.fit_predict(data)

# lazo sobre as equiquetas unicas do array e ordenadas (np.unique)
for label in np.unique(labels):
    #almacenamos os path de todas as imaxes que son asocuiadas coa ultima etiqueta
    labelPaths = imagePaths[np.where(labels == label)]
    
    # lazo sobre todas as imaxes que pertencen a ultima etiqueta
    plt.figure(figsize=[15,15])
    for (i, path) in enumerate(labelPaths):
        # cargamos a imaxe as as visualizamos
        image = cv2.imread(path)
        plt.subplot(1,5,i+1);plt.imshow(image);plt.title("Cluster {}, Imaxe #{}".format(label + 1, i + 1));
```
