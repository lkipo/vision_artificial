# <font style="color:rgb(50,120,229)">Descriptor SIFT</font> 

O descritor SURF, introducido por Bay et al. no seu artigo do ECCV de 2006, [SURF: Speeded Up Robust Features](../artigos/bay_2006.pdf) ,  en realidade moi semellante ao descritor SIFT, pero ten das vantaxes principais.

A primeira vantaxe  que SURF  mis rpido de calcular que SIFT, polo que  mis axeitado para aplicacins en tempo real.

A segunda vantaxe de SURF  que s ten a metade do tamao do descritor SIFT. Lembre da nosa leccin anterior que SIFT devolve un vector de caractersticas de 128 dim  SURF, pola sa banda, devolve un vector de caractersticas de 64 dim.

Imos discutir o descritor **SURF**.

## <font style="color:rgb(50,120,229)">Como funciona SURF?</font>     
     
Do mesmo xeito que en SIFT, o primeiro paso en SURF  seleccionar a rexin rectangular da imaxe que rodea o punto clave. O tamao exacto da rexin determnase durante a fase de deteccin de puntos clave, que normalmente  o mtodo Fast Hessian:



Na segunda etapa de SURF, recorremos cada un dos puntos clave detectados e dividimos a rexin do punto clave en subreas de 4 x 4, de novo, igual que SIFT:


Non obstante, aqu  onde SIFT e SURF comezan a separarse. Para cada unha destas 4 x 4 subreas, as respostas das wavelets de Haar extrense en 5 x 5 puntos de mostra regularmente espazados.



Para cada xanela, calclase unha resposta de Haar tanto na direccin x como na direccin y. Chamaremos a estas respostas $d_{x}$ e $d_{y}$, respectivamente:


Agora que temos $d_{x}$ e $d_{y}$, pesmolas cun kernel gaussiano (como en SIFT). As respostas que estn mis afastadas do centro do punto clave contribuirn menos ao vector de caractersticas final, mentres que as respostas que estn mis preto do centro do punto clave contribuirn mis ao vector de caractersticas final.

O paso final en SURF  calcular o vector de caractersticas. Para cada unha das 4 x 4 subreas, calculamos:
$v = \{\sum d_{x} \sum d_{y} \sum |d_{x}| \sum |d_{y}| \}$



As das primeiras entradas do vector de caractersticas, $\sum d_{x}$ e $\sum d_{y}$, son simplemente abstraccins das respostas de onda de Haar. Nas das segundas entradas, $\sum |d_{x}|$ e $\sum |d_{y}|$, tomamos o valor absoluto para introducir informacin sobre os signos dos cambios de intensidade.

As, temos 4 x 4 = 16 subreas, cada unha delas devolvendo un vector de 4-dim. Estes vectores de caractersticas con 4-dim son todos concatenados. Polo tanto, 16 x 4 = 64-dim para o vector de caractersticas SURF final.

Por ltimo, o vector de caractersticas de 64 dim resultante est normalizado en L2.

De novo, o vector de caractersticas SURF  mis rpido de calcular que SIFT e a metade do tamao de SIFT, polo que o fai mis adecuado tanto para aplicacins en tempo real como para aplicacins con recursos limitados.


## <font style="color:rgb(50,120,229)">Vexamos estas ideas en accin ...</font>

Igual que fixemos no SIFT, mantemos este exemplo de SURF moi breve.

```python
# importamos paquetes
from __future__ import print_function
import argparse
import cv2
import imutils

args={"image":"../dataset/jp_01.png"}

# cargamos a imaxe e pasamola a gris
image = cv2.imread(args["image"])
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Inicializamos o detector para opencv 3+
# detectamos os puntos de interese e achamos sobre eles os descritores SURF
detector = cv2.xfeatures2d.SURF_create()
(kps, descs) = detector.detectAndCompute(gray, None)

# Visualizamos os puntos detectados e a forma dos vectores de caracteristicas
print("[INFO] # de keypoints detectados: {}".format(len(kps)))
print("[INFO] dimensions do vector de caracteristicas: {}".format(descs.shape))
```
