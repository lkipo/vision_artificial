# <font style="color:rgb(50,120,229)">Lemos a imaxe</font>
OpenCV permite ler distintos tipos de imaxes (JPG, PNG, etc). Podemos cargar imaxes de gris, color e con canle de transparencia. Empregamos a instruccin [**`imread`**](https://docs.opencv.org/4.1.0/d4/da8/group__imgcodecs.html#ga288b8b3da0892bd651fce07b3bbd3a56) que ten a seguinte sintaxe:

### <font style="color:rgb(8,133,37)">Sintaxe da funcin </font>
``` python
retval	=	cv2.imread(	filename[, flags]	)
```

Ten **2 argumentos**:

1. `retval`  a imaxe cargada se todo vai ben. En caso contrario  `None`. Isto pasar se o path ou nome son incorrecto ou a imaxe est corrupta.
2. `Path ao ficheiro imaxe`: pode ser un path **absoluto** ou **relativo**. Este  un argumento **obrigatorio**.
3. `Flags`: Estes falgs son empregado para ler unha imaxe nnun xeito particular (por exemplo escala de grises/cor/cun canle alfa). Este  un argumento **opcional** con valor por defecto `cv2.IMREAD_COLOR` ou `1` o cal carga a imaxe como imaxe en cor.

Antes de proceder con exemplo, vexamos cales son os `flags` dispoibles.

**Flags**
1. **`cv2.IMREAD_GRAYSCALE`** ou **`0`**: Carga a imaxe en escala de grises
2. **`cv2.IMREAD_COLOR`** ou **`1`**: carga a imaxe como de cor. Calquera canle de transparencia sere desbotado.  o valor por defecto.
3. **`cv2.IMREAD_UNCHANGED`** ou **`-1`**: Carga a imaxe coa canle de transparencia (alfa).

# <h1 style="font-size:30px;">2.- Coordenadas da imaxe</h1>

Unha vez que os datos da imaxe se almacenan nunha matriz NumPy 2D, podemos visualizalos empregando a funcin matplotlib `imshow ()`como veremos mis adiante pero hai de cousas que hai que ter en conta sobre a imaxe mostrada:

-O sistema de coordenadas usado para representar os datos dos pxeles da imaxe ten a orixe na esquina superior esquerda da imaxe. O ancho da imaxe mdese ao longo do eixe x e a altura ao longo do eixe y. Cando estes datos se almacenan nunha matriz 2D, a primeira posicin do ndice que representa as filas da matriz corresponde ao eixe y e a segunda posicin corresponde ao eixe x. Este sistema de coordenadas mantn a convencin de que o eixo x representa a direccin horizontal e o eixo y a direccin vertical, pero ao acceder aos datos nunha matriz numPy o primeiro ndice representa a coordenada `y` e o segundo ndice representa a coordenada` x `. Esta indexacin aplcase especificamente s matrices 2D, pero debes ter en conta que moitas funcins en OpenCV (e outras bibliotecas) poden referirse a unha coordenada de pxeles na notacin matemtica estndar `(x, y)`. Por exemplo, como veremos en futuros mdulos, moitas funcins en OpenCV requiren que se especifique unha coordenada de pxel como `punto (x, y)`. Resumindo, cando se accede a coordenadas de pxeles nunha matriz 2P NumPy a indexacin sempre  `[y, x]`, pero hai outras veces nas que as coordenadas de imaxe se especifican como `(x, y)`. 



### <font style = "color:rgb(200,0,0)">NOTA</font>
Mantn en mente que en OpenCV, **size** representa unha tupla de `ancho X alto` ou `#columnas X #filas`. Pero en numpy, o mtodo **shape** retorna o tamao como unha tupla de `altura X anchura`.

### <font style = "color:rgb(200,0,0)">NOTA sobre o indexado (a pesar de correr o risco de ser pesado!)</font>

Como mencionamos anteriormente, as matrices son array de numpy, o primeiro ndice ser o `nmero de fila` e o segundo ser `nmero de columna`. Isto  unha fonte de confusin permanente dado que o natural sera pensar nos pixel en termos de coordenadas `(x,y)` ou `(columna,fila)` non en `(fila,columna)`

Por exemplo, para acceder ao elemento na `4 fila` e na `3 columna`, debemos empregar `img[3,4]`. Pero como veremos na seccin de anotacin sobre imaxes, trataremos con puntos que son representados como `(x,y)` e entn, as coordenadas sern `(4,3)`.

# 5. Redimensionar imaxes e volteo de imaxes


<hr style = "border: none; height: 4px; background-color: # D3D3D3" />

A funcin **`resize ()`** redimensiona a imaxe `src` ata o tamao especificado. O tamao e o tipo dervanse de "src", "dsize", "fx" e "fy".

### <font color = "green"> Sintaxe de funcins </font>

```Python
dst = cv2.resize (src, dsize [, dst [, fx [, fy [, interpolacin]]]])
```

`dst`: imaxe de sada; ten o tamao `dsize` (cando non  cero) ou o tamao calculado a partir de `src.size ()`,`fx` e `fy`; o tipo de `dst`  o mesmo que de` src`.

A funcin ten **2 argumentos necesarios**:
    
1. `src`: imaxe de entrada
2. `dsize`: tamao da imaxe de sada

Os argumentos opcionais que se usan a mido inclen:

1. `fx`: factor de escala ao longo do eixo horizontal
2. `fy`: factor de escala ao longo do eixo vertical

A imaxe de sada ten o tamao `dsize` (cando non  cero) ou o tamao calculado a partir de` src.size () `,` fx` e `fy`; o tipo de `dst`  o mesmo que de` src`. Se se especifican `fx` e `fy`, entn `dsize` debera establecerse en `None` xa que  un argumento obrigatorio.

The function **`resize()`** resizes the image `src` down to or up to the specified size. The size and type are derived from the `src`, `dsize`, `fx`, and `fy`.


### <font color="green">ocumentacin de OpenCV</font>

[**`resize()`**](https://docs.opencv.org/4.5.2/da/d54/group__imgproc__transform.html#ga47a974309e9102f5f08231edc7e7529d)

<hr style="border:none; height: 4px; background-color:#D3D3D3" />

````python
img_cor = cv2.imread('../data/images/checkerboard_color.png', cv2.IMREAD_COLOR)

# Mtodo 1: Redimencionamos a imaxe.
resized_image_2x = cv2.resize(img_cor, None, fx = 2, fy = 2)

# Visualizamos as imaxes.
plt.figure(figsize = [20,5])
plt.subplot(121); plt.imshow(img_cor); plt.title('Orixinal')
plt.subplot(122); plt.imshow(resized_image_2x[:, :, ::-1]); plt.title('Orixinal x 2')
````

````python
# Mtodo 2: Redimencionamos a imaxe a un tamao desexado.
desired_width  = 400
desired_height = 200
dim = (desired_width, desired_height)

# Redimensionamos.
resized_image = cv2.resize(img_cor, dsize = dim, interpolation = cv2.INTER_AREA)

plt.figure(figsize = [20,5])
plt.subplot(121); plt.imshow(img_cor); plt.title('Orixinal')
plt.subplot(122); plt.imshow(resized_image [:, :, ::-1]); plt.title('Orixinal x 2')
````

# 5.1. Voltear imaxes

<hr style = "border: none; height: 4px; background-color: # D3D3D3" />

A funcin **`flip ()`** cambia a matriz dun destes tres xeitos diferentes (os ndices de fila e columna estn baseados en 0):

### <font color = "green"> Sintaxe de funcins </font>
```Python
dst = cv2.flip (src, flipCode [, dst])
```

`dst`: matriz de sada do mesmo tamao e tipo que src.

A funcin ten **2 argumentos necesarios**:
    
1. `src`: imaxe de entrada
2. `flipCode`: unha bandeira para especificar como virar a matriz; 0 significa dar a volta ao eixo x e o valor positivo (por exemplo, 1) significa dar a volta ao eixo y. O valor negativo (por exemplo, -1) significa dar a volta a ambos os eixos.


### <font color = "green"> Documentacin OpenCV </font> 

[**`flip()`**](https://docs.opencv.org/4.5.2/d2/de8/group__core__array.html#gaca7be533e3dac7feb70fc60635adf441)

<hr style="border:none; height: 4px; background-color:#D3D3D3" />


````python
# Volteamos a imaxe das tres maneiras
img_barco = cv2.imread('../data/images/barco.jpg', cv2.IMREAD_COLOR)
img_barco_flipped_horz = cv2.flip(img_barco , 1)
img_barco_flipped_vert = cv2.flip(img_barco , 0)
img_barco_flipped_both = cv2.flip(img_barco , -1)

# isualizamos as imaxes.
plt.figure(figsize = [18, 5])
plt.subplot(141); plt.imshow(img_barco_flipped_horz[:, :, ::-1]); 
plt.title('Volteo horizontal');
plt.subplot(142); plt.imshow(img_barco_flipped_vert[:, :, ::-1]); 
plt.title('Volteo  Vertical');
plt.subplot(143); plt.imshow(img_barco_flipped_both[:, :, ::-1]); 
plt.title('Volteo dobre');
plt.subplot(144); plt.imshow(img_barco[:, :, ::-1]);              
plt.title('Orixinal');
````

Hai vairos espazos de cor que se empregan en procesado de imaxes pero para o almacenamento, o mis popular,  o espazo RGB. 

En OpenCV, a orden das canles R, G e B est **invertida**. i.e. na matriz imaxe, a canle azul (B) est indexada primeiro, seguida pola canle verde (G) e finalmente pola vermella (R). O resultado  o espazo BGR e non RGB (non hai razn obxectivas para isto s un criterio)

Notas algo raro na imaxe de cor visialiada aqui? Isto  porque OpenCV emprega formato **BGR** por defecto mentres que Matplotlib asume que a imaxe ser **RGB**. Isto pode ser correxido se: 

* convertemos a imaxe ao espazpo RGB empregando **`cv2.cvtColor(img, cv2.COLOR_BGR2RGB)`** ou 
* invertindo a orden das canles - **`plt.imshow(img[:,:,::-1])`** intercambia a orden das 1 e 3 canle.

````python
# Convertemos o espazo de cor BGR a RGB 
imgRGB = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(imgRGB)
````

````python
# Tamn podemos empregar a seguinte instruccin - Invertir a orden das canles BGR a RGB
plt.imshow(img[:,:,::-1])
````

# Divisin e fusin de canles de cor

Aprendamos agora a dividir e combinar canles de cor usando as funcins ** `split ()` ** e ** `merge ()` ** en OpenCV.

<hr style = "border: none; height: 4px; background-color: # D3D3D3" />

** `split ()` ** Divide unha matriz multicanle en varias matrices de canle nico.

** `merge ()` ** Combina varias matrices para facer unha nica matriz multicanle. Todas as matrices de entrada deben ter o mesmo tamao. 

Vexamos como se pode facer isto.

### <font color="green">Documentacin en OpenCV</font>

[**`split()`**](https://docs.opencv.org/4.5.2/d2/de8/group__core__array.html#ga0547c7fed86152d7e9d0096029c8518a)
[**`merge()`**](https://docs.opencv.org/4.5.2/d2/de8/group__core__array.html#ga61f2f2bde4a0a0154b2333ea504fab1d)

<hr style="border:none; height: 4px; background-color:#D3D3D3" />

````python
# Separa a as canles B,G,R da imaxe
b,g,r = cv2.split(img)

# Visualiza as canles
plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(b);plt.title("Canle B")
plt.subplot(142);plt.imshow(g);plt.title("Canle G")
plt.subplot(143);plt.imshow(r);plt.title("Canle R")

# Unimos as canles BGR da imaxe
imgMerged = cv2.merge((b,g,r))

# Visualizamos a unin
plt.subplot(144);plt.imshow(imgMerged[:,:,::-1]);plt.title("Canles unidas");
````

# <font style="color:rgb(50,120,229)">7. Captura, visualizacin e escritura de vdeo</font>
Un video  unha secuencia de imaxes que vai variando co tempo. O xeito de almacenalos  nun array de catro dimensins (R,G,B,t). Igual que cando tratabamos con imaxes e empregamos a funcin [**`cv2.imread`**](https://docs.opencv.org/4.1.0/d4/da8/group__imgcodecs.html#ga288b8b3da0892bd651fce07b3bbd3a56) de OpenCV para ler unha imaxe almacenada no noso disco duro, esta librara tamn nos aporta unha funcin  **`cv2.VideoCapture`** para criar un [**VideoCapture**](https://docs.opencv.org/4.1.0/d8/dfe/classcv_1_1VideoCapture.html#ac4107fb146a762454a8a87715d9b7c96) que crea un obxecto que nos permite ler un vdeo dende un ficheiro indicado.

### <font style="color:rgb(8,133,37)">Sintaxe da funcin </font>

``` python
<VideoCapture object>	=	cv2.VideoCapture(		)
<VideoCapture object>	=	cv2.VideoCapture(	filename[, apiPreference]	)
<VideoCapture object>	=	cv2.VideoCapture(	index[, apiPreference]	)

```

**Parmetros**

- **`filename`** pode ser:
    - nome dun ficheiro de vdeo (ie. video.avi)
    - ou unha secuencia de imaxes (ie. img_%02d.jpg, lera todos os ficheiros co nomes como img_00.jpg, img_01.jpg, img_02.jpg, ...)
    -ou URL do video stream (ie. protocolo://host:port/script_name?script_params|auth). Nota que cada video stream ou camara IP ten o seu propio esquema URL. Consulta a documentacin correspondente para introducir a URL correcta.
- **`apiPreference`**:	API preferida de captura que queremos empregar.Pode empregar para forzar o uso dun lector especfico se hai varios dispoibles, por exemplo: cv::CAP_FFMPEG ou cv::CAP_IMAGES ou cv::CAP_DSHOW.

```python
cap = cv2.VideoCapture(args)
```

## <font style="color:rgb(50,120,229)">Visualizano os frames como un video [ para local s! ]</font>

Como no caso dunha imaxe, empregaremos [**`cv2.waitKey()`**](https://docs.opencv.org/4.1.0/d7/dfc/group__highgui.html#ga5628525ad33f52eab17feebcfba38bd7) despois da funcin [**`cv2.imshow()`**](https://docs.opencv.org/4.1.0/d7/dfc/group__highgui.html#ga453d42fe4cb60e5723281a89973ee563) para pausar cada frame de vdeo. No caso dunha imaxe, pasamos un  **`0`**  funcin `waitKey`, posto que pausara un tempo indefinido a visualizacin dun frame. Como un vdeo debe dar sensacin de continuidade, precisamos pasar un nmero razonablemente maior que  `0` a waitKey(). **Este nmero  o tempo, en milisegundos, que desexamos visualizar cada frame.**

#### <font style="color:rgb(50,120,229)">Valor de waitKey para Webcam</font>
Cando estamos lendo frame dende a webcam, empregaremos `waitKey(1)` porque as o frame rate da visualizacin estar limitado polo frame rate da cmara.

#### <font style="color:rgb(50,120,229)">Valor de waitKey para un ficheiro de video</font>
Mentre que estamos lendo un vdo que estamos a procesar, pode ser apropiado asignar un retardo de 1 ms a hebra correspondente e permitir o procesado desesado.

**Escribe un script  `videoRead.py` e comproba o funcionamento deste parmetros.**

````python
```python
# Completamos a lectura do video
while(cap.isOpened()):
  # Capturamos frame-a-frame
  ret, frame = cap.read()
    
  if ret == True:
    cv2.imshow("Video", frame)    
    # Esperamos 25 ms antes de movernos ao seguinte frame
    # Isto enlentecera a visualizacion do video
    cv2.waitKey(25)
    
  # Rompemos o lazo
  else: 
    break
```
````

A continuacin veremos como extraer algunha propiedade do vdeo e como cambialas empregando  **`cap.get(propId)`** e **`cap.set(propId,value)`** respectivamente. 

Aqui, **`cap`**  o obxecto **`VideoCapture`** de onde queremos extraer (ou asignar) as propiedades, identificadas por **`propId`** [Property ID](https://docs.opencv.org/4.1.0/d4/d15/group__videoio__flags__base.html#gaeb8dd9c89c10a5c63c139bf7c4f5704d) e **`value`**  o valor que queremos asignarlle  propiedade identificada con id dado por **`propId`**.

Este cadro resume as propiedades mis habituais ID:

| Enumerator | Numerical Value | Property |
| --- | --- | --- |
| **`cv2.CAP_PROP_POS_MSEC`** | `0` | Current position of the video file in milliseconds |
| **`cv2.CAP_PROP_FRAME_WIDTH`** | `3` | Width of the frames in the video stream |
| **`cv2.CAP_PROP_FRAME_HEIGHT`** | `4` | Height of the frames in the video stream |
| **`cv2.CAP_PROP_FPS`** | `5` | Frame rate |
| **`cv2.CAP_PROP_FOURCC`** | `6` | 4-character code of codec |

Profundicemos nisto coa axuda dun exemplo.
