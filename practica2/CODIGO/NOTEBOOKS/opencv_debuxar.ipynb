### Función para visualizar imaxes en Jupyter Notebooks/Google Colab

````python
def plt_imshow(title, image):
        # convertemos a imaxe dende BGR ao espazo de cor RGB e a visualizamos
    if len(image.shape) == 3:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    #plt.imshow(image, cmap="gray")
    plt.imshow(image, cmap="jet")
    plt.title(title)
    plt.grid(False)
    plt.show()
````

### IImplementación das funcións base de debuxo con OpenCV

````python
# Inicializamos o canvas cun mensura de 300x300 pixels e con 3 canles
# (R,G,B) con fondo negro
canvas = np.zeros((300, 300, 3), dtype="uint8")
````

# 2. Debuxando liñas

<hr style = "border: none; height: 4px; background-color: # D3D3D3" />

A función **`line ()`** en OpenCV úsase para debuxar liñas nas imaxes.

### <font color = "green"> Sintaxe da funcióna </font>
"Python
    img = cv2.line (img, pt1, pt2, cor [, espesor [, LineType [, shift]]])
""

**`img`** : a imaxe de saída que foi anotada.

A función ten ** 4 argumentos necesarios **:

1. `img`: Imaxe sobre a que trazaremos unha liña
2. `pt1`: primeiro punto (x, y) localización do segmento de liña
3. `pt2`: segundo punto do segmento de liña
4. `color`: cor da liña que se trazará
    
Os argumentos opcionais que se usan habitualmente inclúen:

1. `espesor`: número enteiro que especifica o grosor da liña. O valor predeterminado é 1.
2. `lineType`: tipo de liña. O valor predeterminado é ** `cv2.LINE_8` **, que significa unha liña de 8 conexións. Normalmente, **`cv2.LINE_AA`** (liña suave) úsase para o lineType.
    
### <font color = "green"> Documentación OpenCV </font>

[**`line()`**](https://docs.opencv.org/4.5.2/d6/d6e/group__imgproc__draw.html#ga7078a9fae8c7e7d13d24dac2520ae4a2)
[**`linetype`**](https://docs.opencv.org/4.5.2/d6/d6e/group__imgproc__draw.html#gaf076ef45de481ac96e0ab3dc2c29a777)

<hr style="border:none; height: 4px; background-color:#D3D3D3" />

````python
# debuxamos unha liña verde dende a esquina superior esquerda do noso canvas
# ata a esquina inferior dereita
green = (0, 255, 0)
cv2.line(canvas, (0, 0), (300, 300), green)
plt_imshow("Canvas", canvas)

# debuxamos unha liña de 3 pixels de gordo dende a esquena superior dereita ata a inferior esquerda
red = (0, 0, 255)
cv2.line(canvas, (300, 0), (0, 300), red, 3)
plt_imshow("Canvas", canvas)
````

# Debuxar un rectángulo

<hr style = "border: none; height: 4px; background-color: # D3D3D3" />

A función **`rectangle ()`** en OpenCV úsase para debuxar rectángulos en imaxes.

### <font color = "green"> Sintaxe de funcións </font>

    img = cv2.rectángulo (img, pt1, pt2, cor [, espesor [, tipo de liña [, desprazamento]]])

`img`: a imaxe de saída que foi anotada.

A función ten **4 argumentos necesarios**:

1. `img`: imaxe sobre a que se debuxa o rectángulo.
2. `pt1`: vértice do rectángulo. Normalmente usamos o **vértice superior esquerdo** aquí.
3. `pt2`: vértice do rectángulo oposto a pt1. Normalmente aquí usamos o vértice **inferior dereito**.
4. `color`: cor do rectángulo
    
Os argumentos opcionais que se usan habitualmente inclúen:

1. "espesor": espesor do contorno do círculo (se é positivo).
    Se se proporciona un valor negativo para este argumento, resultará un rectángulo cheo.
2. `lineType`: tipo do límite do círculo. Isto é exactamente igual ao argumento lineType en
    **`cv2.line ()`**
    
### <font color = "green"> Ligazóns de documentación OpenCV </font>

[**`rectangle()`**](https://docs.opencv.org/4.5.2/d6/d6e/group__imgproc__draw.html#ga07d2f74cadcf8e305e810ce8eed13bc9)

<hr style="border:none; height: 4px; background-color:#D3D3D3" />

````python
# debuxamos un cadrado verde de 50x50 pixel, empezando en 10x10 e rematando en 60x60
cv2.rectangle(canvas, (10, 10), (60, 60), green)
plt_imshow("Canvas", canvas)

# debuxamos outra rectángulo, este vermello con 5 pixels de gordo
cv2.rectangle(canvas, (50, 200), (200, 225), red, 5)
plt_imshow("Canvas", canvas)

# debuxamos un rectángulo (azul e recheo!)
blue = (255, 0, 0)
cv2.rectangle(canvas, (200, 50), (225, 125), blue, -1)
plt_imshow("Canvas", canvas)
````

# Debuxar un círculo

<hr style = "border: none; height: 4px; background-color: # D3D3D3" />

A función **`circle ()`** en OpenCV úsase para debuxar círculos en imaxes.

### <font color = "green"> Sintaxe de funcións </font>

```Python
img = cv2.circle (img, centro, radio, cor [, espesor [, tipo de liña [, desprazamento]]])
```

`img`: a imaxe de saída que foi anotada.

A función ten **4 argumentos necesarios**:

1. `img`: Imaxe sobre a que trazaremos unha liña
2. `center`: centro do círculo
3. `radio`: radio do círculo
4. `cor`: cor do círculo que se debuxará
    
Os argumentos opcionais que se usan habitualmente inclúen:

1. `espesor`: espesor do contorno do círculo (se é positivo).
Se se proporciona un valor negativo para este argumento, resultará un círculo recheo.
2. `lineType`: tipo do límite do círculo. Isto é exactamente igual ao argumento lineType en **`cv2.line ()`**
    
### <font color = "green"> Documentación OpenCV </font>

[**`circle()`**](https://docs.opencv.org/4.5.2/d6/d6e/group__imgproc__draw.html#gaf10604b069374903dbd0f0488cb43670)

<hr style="border:none; height: 4px; background-color:#D3D3D3" />


````python
# reinicializamos o noso canvas como un array baleiro e achamos as coordenadas (x,y) do centro 
canvas = np.zeros((300, 300, 3), dtype="uint8")
(centerX, centerY) = (canvas.shape[1] // 2, canvas.shape[0] // 2)
white = (255, 255, 255)

# lazo no que incrementamos o radio, dende 25 pixels a 150 pixels incrementos de 25
for r in range(0, 175, 25):
	# debusamos un círculo branco co último valor do radio
	cv2.circle(canvas, (centerX, centerY), r, white)

# amosamos a nosa obra de arte!
plt_imshow("Canvas", canvas)
````

````python
# reinicialozamos o noso canvas de novo
canvas = np.zeros((300, 300, 3), dtype="uint8")

# Debuxemos 25 círculos aleatorios
for i in range(0, 25):
	# xeramos aleatoriamente un radio entre 5 e 200 pixels, unha cor aleatoria,
	# e un punto, tamén aleatorio, para ser o centro do círculo que imos debuxar
	radius = np.random.randint(5, high=200)
	color = np.random.randint(0, high=256, size=(3,)).tolist()
	pt = np.random.randint(0, high=300, size=(2,))

	# debuxamos o círculo cos valores aleatorios
	cv2.circle(canvas, tuple(pt), radius, color, -1)

# visualizamos a nosa obra maestra
plt_imshow("Canvas", canvas)
````

# Engadindo texto

<hr style = "border: none; height: 4px; background-color: # D3D3D3" />

A función **`putText ()`** en OpenCV úsase para engadir texto ás imaxes.

### <font color = "green"> Sintaxe de funcións </font>
    img = cv2.putText (img, text, org, fontFace, fontScale, color [, espesor [, lineType [, bottomLeftOrigin]]])

`img`: a imaxe de saída que foi anotada.

A función ten **6 argumentos necesarios**:

1. `img`: imaxe na que hai que escribir o texto.
2. `texto`: cadea de texto a escribir.
3. `org`: esquina inferior esquerda da cadea de texto na imaxe.
4. `fontFace`: tipo de letra
5. `fontScale`: factor de escala de tipo de letra que se multiplica polo tamaño base específico da fonte.
6. `color`: cor do tipo de letra
 
Os argumentos opcionais que se usan habitualmente inclúen:

1. `espesor`: número enteiro que especifica o grosor da liña para o texto. O valor predeterminado é 1.
2. `lineType`: tipo de liña. O valor predeterminado é 8, que significa unha liña conectada a 8. Normalmente, **`cv2.LINE_AA` ** (liña antialiased ou suave) úsase para o lineType.

### <font color = "green"> Documentación OpenCV </font>

[**`putText()`**](https://docs.opencv.org/4.5.2/d6/d6e/group__imgproc__draw.html#ga5126f47f883d730f633d74f07456c576)

<hr style="border:none; height: 4px; background-color:#D3D3D3" />

### Debuxando e escribindo texto sobre imaxes

````python
#empregamos un diccionario en vez de argumentos no Jupyter
# para definir os parámetros de entrada ao noso programa
args = {
	"image": "../data/images/beach.png"
}
````

````python
# cargamos a imaxe
image = cv2.imread(args["image"])

# debuxamos un circulo, dous circulos recheos e un rectánculo a modo emoticona
cv2.circle(image, (168, 188), 90, (0, 0, 255), 2)
cv2.circle(image, (150, 164), 10, (0, 0, 255), -1)
cv2.circle(image, (192, 174), 10, (0, 0, 255), -1)
cv2.rectangle(image, (134, 200), (186, 218), (0, 0, 255), -1)

#escribimos un texto na imaxe
cv2.putText(
     image, #imaxe onde escribimos (array numpy)
     "Ola onda!", #texto
     (30,30), #posicion na que empeza a escritura
     cv2.FONT_HERSHEY_SIMPLEX, #famila da fonte
     1, #tamaño da fonte en puntos
     (0, 0, 255), #cor da fonte
     2) #anchura da fonte


# visualizamos
plt_imshow("Output", image)
````
